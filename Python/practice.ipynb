```python
# Sample function producing a generator object.
# generator object is produced using an yield statement.
# generator objects are memory friendly. We don't have to generate all the numbers and store them in memory.
# As and when required, generator returns the next value in the series
def fib(n):
    a = 0
    b = 1
    while n > 0:
        yield(a)
        n = n - 1
        a, b = b, a+b

def fib_nth_ele(n):
    if n <= 0: return None
    a = 0
    b = 1
    while n > 1:
        a, b = b, a+b
        n = n - 1
    return a
```

```python
# Decorators
# Decorator functions take function as input and always return a function as output
def decor(fn):
    def ret_decor():
        print('Execute functionality before executing normal func')
        fn()
        print('Execute functionality after executing normal func')
    return ret_decor

@decor
@decor
def normal_func():
    print('In Normal Func')

def decor1(fn):
    def ret_decor():
        print('Execute functionality before executing normal func1')
        #fn()
        print('Execute functionality after executing normal func1')
    return ret_decor

@decor1
@decor1
def normal_func1():
    print('In Normal Func1')

normal_func()
normal_func1() # interesting one ... see the commented out fn()
```

```python
# DECORATORS WITH PARAMETERS
# Without Decorator, function math returns 'SUM'
# With Decorator, function math returns list of [SUM, DIFF, MUL, DIV]
def decor(func):
    def decor_func(c,d):
        diff = c - d
        su = func(c,d)
        mul = c * d
        div = c / d
        return [su, diff, mul, div]
    return decor_func

@decor
@decor
def math(a,b):
    return a+b

l = math(2,3)
print(l)
```

```python
# Polymorphism
class A():
    def __init__(self, name):
        self.name = name
    def __len__(self):
        return len(self.name)

class B():
    def __init__(self, age):
        self.age = age
    def __len__(self):
        return self.age

# This is polymorphism as the same signature is used to handle 
# multiple objects in a different way
def length(obj):
    return obj.__len__()

a1 = A('ganesh lalithamaba kvsr sudha varenya')
b1 = B(23)
print(length(a1))
print(length(b1))
print(len(b1))
print(len(a1))
print(A.__len__(a1)) # This is equal to a1.__len__()
print(B.__len__(b1)) # This is equal to b1.__len__()
```

```python
# producer - consumer problem solution with threading
import threading
import queue
import time

class mythread(threading.Thread):
    def __init__(self, name):
        super().__init__()
        self.name = name
        print('Thread {} Created'.format(self.name))
    def run(self):
        print('Thread {} Started'.format(self.name))
        consume_items(self.name)
        print('Thread {} Done'.format(self.name))

def consume_items(name):
    while not exit_flag:
        lck.acquire()
        if not q.empty():
            item = q.get()
            print('Thread {} consumed {}'.format(name, item))
        lck.release()
        time.sleep(1)

exit_flag = 0
lck = threading.Lock()
q = queue.Queue(10)

t1 = mythread('Thread-1')
t2 = mythread('Thread-2')
t3 = mythread('Thread-3')

t1.start()
t2.start()
t3.start()

lck.acquire()
for item in range(90,100):
    q.put(item)
lck.release()

while not q.empty():
    pass

exit_flag = 1
t1.join()
t2.join()
t3.join()

print('End of the Main Program')
```

```python
import threading
import time
import random

class mythread(threading.Thread):
    def __init__(self, name):
        super().__init__()
        self.name = name
        print('{} Initialized'.format(self.name))
    
    def run(self):
        global exit_flag
        if self.name == 'producer':
            cnt = 1
            while not exit_flag:
                lck.acquire()
                produce_item(self.name, cnt)
                cnt += 1
                if cnt == 11: exit_flag = 1
                lck.release()
                time.sleep(random.randint(1,2))
        else:
            while not exit_flag or not q.empty():
                lck.acquire()
                consume_item(self.name)
                lck.release()
                time.sleep(random.randint(1,2))

def produce_item(name, cnt):
    global q
    if q.full(): return
    q.put(cnt)
    print('{} produced item {}'.format(name, cnt))

def consume_item(name):
    global q
    if q.empty(): return
    tmp = q.get()
    print('{} consumed item {}'.format(name, tmp))

lck = threading.Lock()
exit_flag = 0
from queue import Queue
q = Queue(10)
p = mythread('producer')
c1 = mythread('consumer1')
c2 = mythread('consumer2')
p.start()
c1.start()
c2.start()
p.join()
c1.join()
c2.join()
print('End of the Main Program')
```

```python
# Stack with LL Concept
class node():
    def __init__(self, data = None, nxt = None):
        self.data = data
        self.nxt = nxt

class stack():
    def __init__(self):
        self.head = None
    
    def push(self, data):
        tmp = node(data)
        tmp.nxt = self.head
        self.head = tmp
    
    def pop(self):
        if self.head == None:
            return None
        else:
            tmp = self.head.data
            self.head = self.head.nxt
            return tmp
    
    def top(self):
        if self.head == None:
            return None
        else:
            return self.head.data
    
    def is_empty(self):
        return self.head == None

s = stack()
print(s.is_empty())
for item in range(0,100,10):
    s.push(item)
while(not s.is_empty()):
    print(s.pop(), end = ' ')
```

```python
# Queue with LL Concept
class node():
    def __init__(self, data = None, nxt = None):
        self.data = data
        self.nxt = nxt

class que():
    def __init__(self):
        self.front = None
        self.rear = None

    def put(self, data):
        tmp = node(data)
        if self.front == None and self.rear == None:
            self.front = self.rear = tmp
        else:
            self.rear.nxt = tmp
            self.rear = tmp
    
    def get(self):
        if self.front == None and self.rear == None:
            return None
        else:
            tmp = self.front.data
            if self.front == self.rear:
                self.front = self.rear = None
            else:
                self.front = self.front.nxt
            return tmp

    def is_empty(self):
        return self.front == None and self.rear == None

q = que()
print(q.is_empty())
for item in range(0,100,10):
    q.put(item)
print(q.is_empty())
while not q.is_empty():
    print(q.get(), end = ' ')
print()
print(q.is_empty())
```

```python
# Robin-Karp
t = 'abab'
s = 'bacbabababacaca'
def rk(s,t):
    n = len(s)
    m = len(t)
    hs = s[0:m]
    if hs == t:
        return 0
    for i in range(1, n-m+1):
        hs = hs[1:] + s[i+m-1]
        if hs == t:
            return i
    return -1

print(rk(s,t))
```

```python
# Binary Search
def bs(a, low, high, item):
    if low <= high:
        mid = (low + high) // 2
        if a[mid] == item:
            return mid
        elif item < a[mid]:
            return bs(a, low, mid-1, item)
        elif item > a[mid]:
            return bs(a, mid+1, high, item)
    return -1

a = [11,22,33,44,55,100,200,400,500]
print(bs(a, 0, len(a) - 1, 33))
print(bs(a, 0, len(a) - 1, 11))
print(bs(a, 0, len(a) - 1, 500))
print(bs(a, 0, len(a) - 1, 5))
print(bs(a, 0, len(a) - 1, 10000))
```

```python
def gen_ks(k, n):
    if n == 0:
        return []
    elif n == 1:
        return k
    else:
        out = []
        for item in gen_ks(k, n-1):
            for ele in k:
                out.append(ele+item)
        return out

k = ['a', 'b', 'c']
out = gen_ks(k, 4)
print(len(out))
print(out)
```

```python
def comb(l):
    if len(l) <= 1:
        return l
    else:
        out = []
        rv = comb(l[1:])
        out.extend(rv)
        for item in rv:
            out.append(l[0] + item)
        out.append(l[0])
        return out

out = comb(list('abcd'))
print(len(out))
print(out)
```

```python
#AVL
class avlnode():
    def __init__(self, data = None):
        self.data = data
        self.left = None
        self.right = None
        self.height = 0

class avl():
    def __init__(self):
        self.root = None

    def height(self, t):
        if t == None:
            return -1
        else:
            return t.height

    def srotate_left_right(self, t):
        d = t.left
        t.left = d.right
        d.right = t
        t.height = max(self.height(t.left), self.height(t.right)) + 1
        d.height = max(self.height(t), self.height(d.left)) + 1
        return d

    def srotate_right_left(self, t):
        d = t.right
        t.right = d.left
        d.left = t
        t.height = max(self.height(t.left), self.height(t.right)) + 1
        d.height = max(self.height(t), self.height(d.right)) + 1
        return d

    def drotate_left_right(self, t):
        t.left = self.srotate_right_left(t.left)
        return self.srotate_left_right(t)

    def drotate_right_left(self, t):
        t.right = self.srotate_left_right(t.right)
        return self.srotate_right_left(t)

    def inserth(self, t, data):
        if t == None:
            return avlnode(data)
        elif data < t.data:
            t.left = self.inserth(t.left, data)
            if self.height(t.left) - self.height(t.right) == 2:
                if data < t.left.data:
                    return self.srotate_left_right(t)
                elif data > t.left.data:
                    return self.drotate_left_right(t)
        elif data > t.data:
            t.right = self.inserth(t.right, data)
            if self.height(t.right) - self.height(t.left) == 2:
                if data > t.right.data:
                    return self.srotate_right_left(t)
                elif data < t.right.data:
                    return self.drotate_right_left(t)

        t.height = max(self.height(t.left), self.height(t.right)) + 1
        return t

    def insert(self, data):
        self.root = self.inserth(self.root, data)

    def preorderh(self, t):
        if t != None:
            print(t.data, end = ' ')
            self.preorderh(t.left)
            self.preorderh(t.right)

    def preorder(self):
        self.preorderh(self.root)
        print()

    def levelorderh(self, t):
        if t == None:
            return
        from queue import Queue
        q = Queue()
        q.put(t)
        while(not q.empty()):
            tmp = q.get()
            print(tmp.data, end = ' ')
            if tmp.left != None:
                q.put(tmp.left)
            if tmp.right != None:
                q.put(tmp.right)

    def levelorder(self):
        self.levelorderh(self.root)
        print()
        
    def lcah(self, t, p, q):
        if t == None:
            return None
        elif t.data == p or t.data == q:
            return t.data
        elif (p < t.data and q > t.data) or (p > t.data and q < t.data):
            return t.data
        elif p < t.data and q < t.data:
            return self.lcah(t.left, p, q)
        elif p > t.data and q > t.data:
            return self.lcah(t.right, p, q)

    def lca(self, p, q):
        return self.lcah(self.root, p, q)

t1 = avl()
for item in [7,3,5,2,9]:
    t1.insert(item)
t1.preorder()
t1.levelorder()
print(t1.height(t1.root.left))
print(t1.root.data)
print(t1.root.left.data)
print(t1.root.right.data)
print(t1.root.left.left.data)
print(t1.root.right.right.data)
print(t1.root.left.right)
print(t1.root.right.left)
print(t1.lca(2,9))
print(t1.lca(9,5))
print(t1.lca(2,5))
print(t1.lca(3,5))
print(t1.lca(33,22))
```

```python
# Binary Tree
class tnode():
    def __init__(self, data = None):
        self.data = data
        self.left = None
        self.right = None
class btree():
    def __init__(self):
        self.root = None
    
    def inserth(self, t, data):
        if t == None:
            return tnode(data)
        from queue import Queue
        q = Queue()
        q.put(t)
        while(not q.empty()):
            tmp = q.get()
            if tmp.left != None:
                q.put(tmp.left)
            else:
                tmp.left = tnode(data)
                break
            
            if tmp.right != None:
                q.put(tmp.right)
            else:
                tmp.right = tnode(data)
                break

        return t

    def insert(self, data):
        self.root = self.inserth(self.root, data)

    def preorderh(self, t):
        if t != None:
            print(t.data, end = ' ')
            self.preorderh(t.left)
            self.preorderh(t.right)

    def preorder(self):
        self.preorderh(self.root)
        print()

    def levelorderh(self, t):
        if t == None:
            return
        from queue import Queue
        q = Queue()
        q.put(t)
        while(not q.empty()):
            tmp = q.get()
            print(tmp.data, end = ' ')
            if tmp.left != None:
                q.put(tmp.left)
            if tmp.right != None:
                q.put(tmp.right)

    def levelorder(self):
        self.levelorderh(self.root)
        print()

    def findh(self, t, data):
        if t == None:
            return False
        elif data == t.data:
            return True
        else:
            return self.findh(t.left, data) or self.findh(t.right, data)
        
    def find(self, data):
        return self.findh(self.root, data)

    def lcah(self, t, p, q):
        if t == None:
            return None
        elif t.data == p or t.data == q:
            return t.data
        else:
            left = self.lcah(t.left, p, q)
            right = self.lcah(t.right, p, q)
            if left and right:
                return t.data
            else:
                return left if left else right

    def lca(self, p, q):
        return self.lcah(self.root, p, q)

b1 = btree()
for item in [7,3,5,2,9]:
    b1.insert(item)
b1.preorder()
b1.levelorder()
print(b1.find(9))
print(b1.find(99))
print(b1.lca(2,9))
print(b1.lca(9,5))
print(b1.lca(2,5))
print(b1.lca(3,5))
print(b1.lca(33,22))
```

```python
def power(k, n):
    if n == 0: return 1
    x = power(k, n // 2)
    if n % 2 == 1: return k * (x ** 2)
    else: return x ** 2

print(power(2,6))
print(power(2,4))
print(power(2,10))
print(power(2,11))
```

```python
# Matrix Mul
def mat_mul(a,b):
    if not (a and b): return
    p = len(a)
    q = len(a[0])
    r = len(b[0])
    c = [[0 for j in range(r)] for i in range(p)]
    for i in range(p):
        for j in range(r):
            for k in range(q):
                c[i][j] += a[i][k] * b[k][j]
    return c

a = [[1,2],[3,4]]
b = [[5],[7]]
print(mat_mul(a,b))
```

```python
# Singly Linked List
class node():
    def __init__(self, data=None, nxt=None):
        self.data = data
        self.nxt = nxt
    
class sll():
    def __init__(self):
        self.head = None
    
    def insert_at_beg(self, data):
        tmp = node(data)
        tmp.nxt = self.head
        self.head = tmp
    
    def insert(self, pos, data):
        if pos < 0:
            print('Negative Index')
            return
        if pos == 0:
            self.insert_at_beg(data)
        else:
            i = 0
            h = self.head
            while (i < pos - 1):
                h = h.nxt
                i += 1
                if h == None:
                    print('Index out of range')
                    return
            tmp = node(data)
            tmp.nxt = h.nxt
            h.nxt = tmp

    def plist(self):
        tmp = self.head
        while(tmp != None):
            print(tmp.data, end = ' ')
            tmp = tmp.nxt
        print()

    def delete_at_beg(self):
        if self.head == None: return
        self.head = self.head.nxt
    
    def delete(self, pos):
        if pos < 0:
            print('Negative Index Delete')
            return
        if pos == 0:
            self.delete_at_beg()
        else:
            i = 0
            h = self.head
            while (i < pos - 1):
                h = h.nxt
                i += 1
                if h == None:
                    print('Index out of range')
                    return
            if h.nxt == None:
                print('Index out of range')
                return
            h.nxt = h.nxt.nxt

    def remove(self, data):
        if self.head == None: return
        h = self.head
        if h.data == data:
            self.head = self.head.nxt
            return
        while(h.nxt != None and h.nxt.data != data):
            h = h.nxt
        if h.nxt == None:
            return
        h.nxt = h.nxt.nxt
    
    def reverse(self):
        cur = self.head
        prev = None
        while(cur != None):
            n = cur.nxt
            cur.nxt = prev
            prev = cur
            cur = n
        self.head = prev

    def kth_from_last(self, k):
        if self.head == None:
            print('Index out of range - kth from last')
            return
        sp = self.head
        fp = self.head
        i = 0
        while(fp != None and i < k):
            fp = fp.nxt
            i += 1
            if fp == None and i < k:
                print('Index out of range - kth from last')
                return

        while fp != None:
            fp = fp.nxt
            sp = sp.nxt
        
        return sp.data

    def middle_node(self):
        if self.head == None:
            return
        sp = fp = self.head
        while(fp != None):
            fp = fp.nxt
            if fp == None:
                break
            fp = fp.nxt
            if fp == None:
                break
            sp = sp.nxt
        return sp.data

l = sll()
for item in range(5):
    l.insert(0,item)
    l.plist()
l.reverse()
l.plist()
l.insert(5,7)
l.plist()
l.insert(3,20)
l.plist()
l.insert(0,37)
l.plist()
l.insert(-3, 22)
l.plist()
l.reverse()
l.plist()
l.delete(4)
l.plist()
l.delete(7)
l.plist()
l.delete(6)
l.plist()
l.delete(0)
l.plist()
l.remove(22)
l.plist()
l.remove(4)
l.plist()
l.remove(0)
l.plist()
l.remove(2)
l.plist()
l.remove(33)
l.plist()
l.reverse()
l.plist()
print(l.kth_from_last(4))
print(l.middle_node())
l.insert(0, 23)
l.plist()
print(l.middle_node())
l.remove(23)
l.remove(1)
l.plist()
print(l.middle_node())
l.remove(20)
l.plist()
print(l.middle_node())
l.remove(3)
l.plist()
print(l.middle_node())
```

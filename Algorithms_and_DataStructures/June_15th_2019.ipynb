```python
# Longest Common Substring
def longest_common_substring(s, t):
    m = len(s)
    n = len(t)
    d = [[0 for _ in range(n+1)] for _ in range(m+1)]
    maxi = 0
    maxn = 0
    for i in range(1,m+1):
        for j in range(1, n+1):
            if s[i-1] == t[j-1]:
                d[i][j] = 1 + d[i-1][j-1]
                if d[i][j] > maxn:
                    maxn = d[i][j]
                    maxi = i-1
            else:
                d[i][j] = 0
    #print(maxn, maxi)
    return s[maxi-maxn+1:maxi+1]

s = 'ganesh is the pratham pujya'
t = 'g@a#n$e@s@a'
print(longest_common_substring(s,t))
```

```python
def longest_common_subsequence(s, t):
    m = len(s)
    n = len(t)
    d = [[0 for _ in range(n+1)] for _ in range(m+1)]
    for i in range(1,m+1):
        for j in range(1, n+1):
            if s[i-1] == t[j-1]:
                d[i][j] = 1 + d[i-1][j-1]
            else:
                d[i][j] = max(d[i-1][j], d[i][j-1])
    match = ''
    i = m
    j = n
    while i > 0 and j > 0:
        if d[i][j] == d[i-1][j]:
            i = i - 1
        elif d[i][j] == d[i][j-1]:
            j = j - 1
        elif d[i][j] == 1 + d[i-1][j-1]:
            match = s[i-1] + match
            i = i - 1
            j = j - 1

    return match

s = 'ganesa'
t = 'g#a#n#e#s#a#'
print(longest_common_subsequence(s,t))
```

```python
import functools
@functools.lru_cache()
def number_of_coins(d, j):
    if j < 0: return float("inf")
    if j == 0: return 0
    mini = float("inf")
    for item in d:
        mini = min(mini, number_of_coins(d, j-item)+1)
    return mini

d = (1,5,10,20,25)
j = 250
print(number_of_coins(d, j))
```

```python
import functools
@functools.lru_cache()
def integer_knapsack(d, v, j):
    if j < 0: return float("-inf")
    if j == 0: return 0
    maxi = float("-inf")
    for i in range(len(d)):
        maxi = max(maxi, integer_knapsack(d, v, j-d[i])+v[i], integer_knapsack(d, v, j-1))
    return maxi

d = (5,10,20,25)
v = (5,10,50,100)
j = 40
print(integer_knapsack(d, v, j))
```

```python
def sum_of_subsets(w, s):
    k = sum(s)
    if w > k: return (0, [])
    arr = [0 for _ in range(k+1)]
    combi = [[] for _ in range(k+1)]
    arr[0] = 1
    combi[0] = [[]]
    where_to_start = 0
    for item in s:
        for j in range(where_to_start, -1, -1):
            if arr[j] == 1:
                arr[j+item] = 1
                where_to_start = max(where_to_start, j+item)
                for l in combi[j]:
                    combi[j+item].append(l + [item])

    return (arr[w], combi[w])

s = (2,5,10,15,20,25)
w = 25
print(sum_of_subsets(w,s))
```

```python
def zero_one_knapsack(w, v, m):
    n = len(w)
    d = [[0 for _ in range(m+1)] for _ in range(n+1)]

    for i in range(1,n+1):
        for j in range(1,m+1):
            if j >= w[i-1]:
                d[i][j] = max(d[i-1][j], d[i-1][j-w[i-1]] + v[i-1])
            else:
                d[i][j] = d[i-1][j]
    
    return max(d[n])

w = (2,5,10,20,25)
v = (2,5,10,50,100)
j = 59
print(zero_one_knapsack(w, v, j))
```

```python
def sum_of_subsets_matrix(m,w):
    n = len(w)
    d = [[0 for _ in range(m+1)] for _ in range(n+1)]
    for i in range(n+1):
        d[i][0] = 1
    for i in range(1,n+1):
        for j in range(1,m+1):
            if j >= w[i-1]:
                d[i][j] = max(d[i-1][j], d[i-1][j-w[i-1]])
            else:
                d[i][j] = d[i-1][j]
    
    l = []
    i = n
    j = m
    if d[n][m] == 1:
        while i > 0 and j > 0:
            if j - w[i-1] >= 0 and d[i][j] == d[i-1][j-w[i-1]]:
                l.insert(0, w[i-1])
                j = j - w[i-1]
                i = i -1
            else:
                i = i -1

    return (d[n][m], l)

w = (2,5,10,15,20,25)
m = 17
print(sum_of_subsets_matrix(m,w))
```

```python
def max_contiguous_sum(arr):
    maxi = max(arr)
    if maxi < 0: return maxi
    sum_ending_here = 0
    global_sum = 0
    for i in range(len(arr)):
        sum_ending_here = sum_ending_here + arr[i]
        if sum_ending_here < 0:
            sum_ending_here = 0
            continue
        if sum_ending_here > global_sum:
            global_sum = sum_ending_here
    return global_sum

def max_contiguous_sum1(arr):
    maxi = max(arr)
    if maxi < 0: return maxi
    m = [0 for _ in range(len(arr))]
    if arr[0] > 0: m[0] = arr[0]
    else: m[0] = 0
    for i in range(1, len(arr)):
        if m[i-1] + arr[i] > 0:
            m[i] = m[i-1] + arr[i]
        else:
            m[i] = 0
    return max(m)

print(max_contiguous_sum([-2,11,-4,13,-5,2]))
print(max_contiguous_sum([1,-3,4,-2,-1,6]))
print(max_contiguous_sum([-1,-2,-3]))
print(max_contiguous_sum1([-2,11,-4,13,-5,2]))
print(max_contiguous_sum1([1,-3,4,-2,-1,6]))
print(max_contiguous_sum1([-1,-2,-3]))
```

```python
def binary_search_helper(a, low, high, data):
    while low <= high:
        mid = (low + high) // 2
        if a[mid] == data: return mid
        elif data < a[mid]: high = mid - 1
        else: low = mid + 1
    return None

def binary_search(arr, data):
    return binary_search_helper(arr, 0, len(arr) - 1, data)

arr = [-33, -22, -11, 11, 33, 44, 66, 77, 88, 99, 100, 500]
print(binary_search(arr, -33))
```

```python
def exact_string_match(s,t):
    n = len(s)
    m = len(t)
    for i in range(0, n-m+1):
        for j in range(0, m):
            if s[i+j] != t[j]:
                break
        else:
            return i
    return None

s = 'bacbabzyxababacaca'
t = 'ababaca'
t = 'ababaca'
s = 'ababdcababaca'
print(exact_string_match(s,t))
```

```python
from heapq import heapify, heappush, heappop, nsmallest
class median_heap():
    def __init__(self):
        self.maxheap = []
        self.minheap = []
        heapify(self.maxheap)
        heapify(self.minheap)
        self.n = 0
        self.maxn = 0
        self.minn = 0
    
    def insert(self, data):
        if self.n == 0:
            heappush(self.maxheap, data * -1)
            self.n += 1
            self.maxn += 1
        elif self.maxn == self.minn + 1:
            if data > (nsmallest(1, self.maxheap)[0] * -1):
                heappush(self.minheap, data)
                self.n += 1
                self.minn += 1
            else:
                tmp = heappop(self.maxheap) * -1
                heappush(self.minheap, tmp)
                heappush(self.maxheap, data * -1)
                self.n += 1
                self.minn += 1
        elif self.maxn == self.minn:
            if data > (nsmallest(1, self.minheap)[0]):
                tmp = heappop(self.minheap)
                heappush(self.maxheap, tmp * -1)
                heappush(self.minheap, data)
                self.n += 1
                self.maxn += 1
            else:
                heappush(self.maxheap, data * -1)
                self.n += 1
                self.maxn += 1

    def get_median(self):
        if self.n % 2 == 0:
            return ((nsmallest(1, self.maxheap)[0] * -1) + (nsmallest(1, self.minheap)[0])) / 2
        else:
            return nsmallest(1, self.maxheap)[0] * -1

m = median_heap()
for item in [33,11,77,22,99,-100,100,1000]:
    m.insert(item)
    print(m.maxheap, m.minheap, m.get_median())
    #print(m.get_median())
```

```python
def perm(a, i, n, res):
    if i == n:
        res.append(''.join(a))
    else:
        for k in range(i,n):
            a[i], a[k] = a[k], a[i]
            perm(a, i+1, n, res)
            a[i], a[k] = a[k], a[i]

res = []
perm(list('abc'), 0, len('abc'), res)
print(len(res))
print(res)
```

```python
def comb_helper(a, n):
    if n == 0: return []
    elif n == 1: return a
    else:
        res = []
        tmp = comb_helper(a[1:], n-1)
        res.extend(tmp)
        for item in tmp:
            res.append(a[0] + item)
        res.append(a[0])
        return res

def comb(a, k):
    return list(filter(lambda x: len(x) == k, comb_helper(a, len(a))))

s = 'abc'
print(comb(list(s), 2))
```

```python
def kth_smallest_in_2_sorted_arrays_helper(a, b, k, n, m):
    if k > m+n or k < 1: return None
    if n > m:
        a,b = b,a
        n,m = m,n
    stepsa = (min(n,k) - 1) // 2
    stepsb = k - stepsa - 2
    if stepsb > m-1: stepsb = m-1
    if n == 0: return b[k-1]
    elif k == 1: return min(a[0], b[0])
    elif a[stepsa] == b[stepsb]: return a[stepsa]
    elif a[stepsa] < b[stepsb]: return kth_smallest_in_2_sorted_arrays_helper(a[stepsa+1:n], b, k-stepsa-1, n-stepsa-1, m)
    else: return kth_smallest_in_2_sorted_arrays_helper(a, b[stepsb+1:m], k-stepsb-1, n, m-stepsb-1)

def kth_smallest_in_2_sorted_arrays(a, b, k):
    n = len(a)
    m = len(b)
    return kth_smallest_in_2_sorted_arrays_helper(a, b, k, n, m)

a = [-200, -100, -66, 3, 77]
b = [-20, 20, 67]
c = a + b
c.sort()
print(c)
k = 8
print(kth_smallest_in_2_sorted_arrays(a, b, k))
```

```python
def edit_distance_helper(s,t,m,n):
    d = [[0 for _ in range(n+1)] for _ in range(m+1)]
    for i in range(m+1): d[i][0] = i
    for j in range(n+1): d[0][j] = j
    for i in range(1,m+1):
        for j in range(1,n+1):
            cost = 1
            if s[i-1] == t[j-1]: cost = 0
            d[i][j] = min(d[i-1][j] + 1, d[i][j-1] + 1, d[i-1][j-1] + cost)
    return d[m][n]

def edit_distance(s,t):
    return edit_distance_helper(s,t,len(s),len(t))

s = 'ganesa'
t = 'asaz'
print(edit_distance(s,t))
```

```python
def lps(s):
    n = len(s)
    d = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n): d[i][i] = 1
    for i in range(n-1,-1,-1):
        for j in range(n):
            if i < j:
                if s[i] == s[j]:
                    if i+1 <= j-1:
                        d[i][j] = 2 + d[i+1][j-1]
                    else:
                        d[i][j] = 2
                else:
                    d[i][j] = max(d[i+1][j], d[i][j-1])
    match = []
    i = 0
    j = n-1
    while i <= j:
        if d[i][j] == d[i+1][j]:
            i = i + 1
        elif d[i][j] == d[i][j-1]:
            j = j - 1
        elif d[i][j] == 2 + d[i+1][j-1]:
            tmp = len(match) // 2
            match.insert(tmp, s[i])
            match.insert(tmp, s[i])
            i = i + 1
            j = j - 1
        elif d[i][j] == 1 + d[i+1][j-1]:
            tmp = len(match) // 2
            match.insert(tmp, s[i])
            i = i + 1
            j = j - 1
    return (d[0][n-1], ''.join(match))

print(lps('asddta'))
```

```python
def lpss(s):
    n = len(s)
    d = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        d[i][i] = 1
    maxi = 0
    maxj = 0
    for i in range(n-1, -1, -1):
        for j in range(n):
            if i < j:
                if s[i] == s[j]:
                    if i+1 <= j-1:
                        d[i][j] = d[i+1][j-1]
                    else:
                        d[i][j] = 1
                    if d[i][j] == 1:
                        if j-i > maxj-maxi:
                            maxi = i
                            maxj = j
                else:
                    d[i][j] = 0
    return (s[maxi:maxj+1])

print(lpss('asdsa'))
```

```python
nneg = 0
h = dict()

def decr(ch):
    global h
    global nneg
    if ch not in h: return
    if h[ch] == 0:
        nneg = nneg - 1
    h[ch] = h[ch] - 1

def incr(ch):
    global h
    global nneg
    if ch not in h: return
    h[ch] += 1
    if h[ch] == 0:
        nneg += 1

def min_window(s,p):
    for ch in p:
        h[ch] = 0

    for ch in p:
        decr(ch)

    print(nneg, h)

    n = len(s)
    j = 0
    i = 0
    mini = float("inf")
    minj = float("inf")
    while i < n:
        while nneg >= 0 and j < n:
            if mini == float("inf") or (i - j) < (mini - minj):
                mini = i
                minj = j
            #print(i,j,mini,minj)
            decr(s[j])
            j += 1            
        incr(s[i])
        i += 1
    if mini == float("inf"): return None
    else: return s[minj:mini]

p = 'abcda'
s = 'adcfbdacgbzw'
print(min_window(s,p))
```

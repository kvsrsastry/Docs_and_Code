```python
def peak_helper(a, low, high):
    if low > high: return None
    while (low + 1) < high: # minimum 3 elements needed
        mid = (low + high) // 2
        if a[mid-1] <= a[mid] >= a[mid+1]:
            return mid
        elif a[mid-1] > a[mid]:
            high = mid
        elif a[mid+1] > a[mid]:
            low = mid
    
    if a[low] > a[high]: return low
    else: return high

def peak(a):
    return peak_helper(a, 0, len(a) - 1)

a = [11, 22, 3, 144, 550, -66, 77]
print(peak(a))
```

```python
def shell_sort(a):
    n = len(a)
    h = n // 2
    while h > 0:
        for i in range(1, n):
            j = i
            tmp = a[i]
            while j-h >= 0 and a[j-h] > tmp:
                a[j] = a[j-h]
                j = j - h
            a[j] = tmp
        h = h // 2

a = [3, 144, 550, -66, 77, 1023, -100, -200]
shell_sort(a)
print(a)
```

```python
class disjoint_set():
    def __init__(self, nodes):
        self.d = dict()
        for v in nodes:
            self.d[v] = -1
    
    def find(self, v):
        while type(self.d[v]) is not int:
            v = self.d[v]
        return v
    
    def union(self, v1, v2):
        root1 = self.find(v1)
        root2 = self.find(v2)
        if root1 != root2:
            if self.d[root1] < self.d[root2]:
                self.d[root1] += self.d[root2]
                self.d[root2] = root1
            else:
                self.d[root2] += self.d[root1]
                self.d[root1] = root2
```

```python
class node():
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

class stack():
    def __init__(self):
        self. head = None
    
    def push(self, data):
        tmp = node(data)
        tmp.next = self.head
        self.head = tmp
    
    def pop(self):
        if self.head == None:
            raise Exception('Stack Empty ...')
        else:
            tmp = self.head
            self.head = self.head.next
            tmp_data = tmp.data
            del tmp
            return tmp_data
    
    def top(self):
        if self.head == None:
            raise Exception('Stack Empty ...')
        else:
            return self.head.data
    
    def empty(self):
        return self.head == None

s = stack()
print(s.empty())
for item in range(20,27):
    s.push(item)
print(s.empty())
print(s.top())
while not s.empty():
    print(s.pop(), end = ' ')
print()
print(s.empty())
```

```python
class queue():
    def __init__(self):
        self.front = None
        self.rear = None
    
    def put(self, data):
        tmp = node(data)
        if self.front == None and self.rear == None:
            self.front = self.rear = tmp
        else:
            self.rear.next = tmp
            self.rear = tmp
    
    def get(self):
        if self.front == None and self.rear == None:
            raise Exception('Queue Empty ...')
        else:
            tmp = self.front
            tmp_data = self.front.data
            if self.front == self.rear:
                self.front = self.rear = None
            else:
                self.front = self.front.next
            del tmp
            return tmp_data

    def empty(self):
        return self.front == None

q = queue()
print(q.empty())
for item in range(20, 27):
    q.put(item)
print(q.empty())
while not q.empty():
    print(q.get(), end = ' ')
print()
print(q.empty())
print(q.get())
```

```python
def fib_seq(n):
    a, b = 0, 1
    while n > 0:
        print(a, end = ' ')
        a, b = b, a+b
        n -= 1
    print()

def nth_fib(n):
    a, b = 0, 1
    while n > 1:
        a, b = b, a+b
        n = n - 1
    print(a)

fib_seq(8)
nth_fib(9)
```

```python
def mcs(a):
    n = len(a)
    if max(a) < 0: return max(a)
    sum_ending_here = 0
    global_sum = 0
    for i in range(n):
        sum_ending_here += a[i]
        if sum_ending_here < 0:
            sum_ending_here = 0
            continue
        if sum_ending_here > global_sum:
            global_sum = sum_ending_here

    return global_sum

def mcs1(a):
    m = [0 for _ in range(len(a))]
    m[0] = 0 if a[0] < 0 else a[0]
    for i in range(1,len(a)):
        if m[i-1] + a[i] > 0:
            m[i] = m[i-1] + a[i]
        else:
            m[i] = 0
    
    maxi = float("-inf")
    for i in range(len(a)):
        if m[i] > maxi: maxi = m[i]
    return maxi

print(mcs([-2,11,-4,13,-5,2]))
print(mcs([1,-3,4,-2,-1,6]))
print(mcs([-1,-2,-3]))
```

```python
def opt_mat_mul(p):
    n = len(p) - 1
    return opt_mat_mul_helper(p, 1, n)

def opt_mat_mul_helper(p, i, j):
    if i == j: return 0
    mini = float("inf")
    for k in range(i, j):
        mini = min(mini, opt_mat_mul_helper(p, i, k) + opt_mat_mul_helper(p, k+1, j) + p[i-1]*p[k]*p[j])
    return mini

p = [5,4,3,2,11,7]
print(opt_mat_mul(p))
```

```python
# Remove Duplicates and 
# Output the first one in the lexicographical order
def dup_lex(a):
    from collections import Counter
    cnter = Counter(a)
    n = len(a)
    i = 1
    while i < n:
        if i > 0 and a[i-1] > a[i] and cnter[a[i-1]] > 1:
            cnter[a[i-1]] -= 1
            del a[i-1]
            n = n - 1
            i = i - 2
        elif i > 0 and a[i-1] > a[i] and cnter[a[i-1]] == 1 and cnter[a[i]] > 1:
            cnter[a[i]] -= 1
            del a[i]
            n = n - 1
            i = i - 1
        i += 1

    return ''.join(a)

a = 'cbacdbcb'
print(dup_lex(list(a)))
```

```python
class avlnode():
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None
        self.height = 0

class avl():
    def __init__(self):
        self.root = None
    
    def findh(self, t, data):
        if t == None:
            return None
        elif data == t.data:
            return t
        elif data < t.data:
            return self.findh(t.left, data)
        elif data > t.data:
            return self.findh(t.right, data)
    
    def find(self, data):
        return self.findh(self.root, data)

    def find_maxh(self, t):
        if t == None: return None
        while t.right != None:
            t = t.right
        return t

    def find_max(self):
        return self.find_maxh(self.root)

    def find_minh(self, t):
        if t == None: return None
        while t.left != None:
            t = t.left
        return t

    def find_min(self):
        return self.find_minh(self.root)

    def deleteh(self, t, data):
        if t == None:
            return None
        elif data < t.data:
            t.left = self.deleteh(t.left, data)
            if self.height(t.right) - self.height(t.left) == 2:
                t = self.single_rotate_left(t)

        elif data > t.data:
            t.right = self.deleteh(t.right, data)
            if self.height(t.left) - self.height(t.right) == 2:
                t = self.single_rotate_right(t)

        elif data == t.data:
            if t.left != None and t.right != None:
                tmp = self.find_minh(t.right)
                t.data = tmp.data
                t.right = self.deleteh(t.right, tmp.data)
            else:
                return t.left if t.left else t.right

        t.height = max(self.height(t.left), self.height(t.right)) + 1
        return t

    def delete(self, data):
        self.root = self.deleteh(self.root, data)

    def height(self, t):
        if t == None: return -1
        else: return t.height

    def single_rotate_right(self, t):
        p = t.left
        t.left = p.right
        p.right = t
        t.height = max(self.height(t.left), self.height(t.right)) + 1
        p.height = max(self.height(p.left), self.height(p.right)) + 1
        return p

    def single_rotate_left(self, t):
        p = t.right
        t.right = p.left
        p.left = t
        t.height = max(self.height(t.left), self.height(t.right)) + 1
        p.height = max(self.height(p.left), self.height(p.right)) + 1
        return p

    def double_rotate_right(self, t):
        t.left = self.single_rotate_left(t.left)
        return single_rotate_right(t)

    def double_rotate_left(self, t):
        t.right = self.single_rotate_right(t.right)
        return self.single_rotate_left(t)

    def inserth(self, t, data):
        if t == None:
            t = avlnode(data)
        elif data < t.data:
            t.left = self.inserth(t.left, data)
            if self.height(t.left) - self.height(t.right) == 2:
                if data < t.left.data:
                    t = self.single_rotate_right(t)
                else:
                    t = self.double_rotate_right(t)
                
        elif data > t.data:
            t.right = self.inserth(t.right, data)
            if self.height(t.right) - self.height(t.left) == 2:
                if data > t.right.data:
                    t = self.single_rotate_left(t)
                else:
                    t = self.double_rotate_left(t)

        t.height = max(self.height(t.left), self.height(t.right)) + 1
        return t

    def insert(self, data):
        self.root = self.inserth(self.root, data)

    def inorderh(self, t):
        if t != None:
            self.inorderh(t.left)
            print(t.data, end = ' ')
            self.inorderh(t.right)

    def inorder(self):
        self.inorderh(self.root)
        print()

    def preorderh(self, t):
        if t != None:
            print(t.data, end = ' ')
            self.preorderh(t.left)
            self.preorderh(t.right)

    def preorder(self):
        self.preorderh(self.root)
        print()

    def postorderh(self, t):
        if t != None:
            self.postorderh(t.left)
            self.postorderh(t.right)
            print(t.data, end = ' ')

    def postorder(self):
        self.postorderh(self.root)
        print()

    def levelorderh(self, t):
        if t == None: return
        from queue import Queue
        q = Queue()
        q.put(t)
        while not q.empty():
            tmp = q.get()
            print(tmp.data, end = ' ')
            if tmp.left != None:
                q.put(tmp.left)
            if tmp.right != None:
                q.put(tmp.right)

    def levelorder(self):
        self.levelorderh(self.root)
        print()

t = avl()
for item in range(5):
    t.insert(item)
t.insert(-1)
t.preorder()
t.inorder()
t.postorder()
t.levelorder()
t.delete(0)
t.preorder()
t.inorder()
t.postorder()
t.levelorder()
t.delete(-1)
t.preorder()
t.inorder()
t.postorder()
t.levelorder()
t.delete(3) # FIX THIS CASE
t.preorder()
t.inorder()
t.postorder()
t.levelorder()
```

```python
class node():
    def __init__(self, data = None, next = None):
        self.data = data
        self.next = next
    
class sll():
    def __init__(self):
        self.head = None

    def insert_at_beg(self, data):
        new = node(data)
        new.next = self.head
        self.head = new

    def insert(self, pos, data):
        if pos < 0: raise Exception('pos < 0')
        elif pos == 0:
            self.insert_at_beg(data)
        else:
            tmp = self.head
            cnt = 0
            while cnt < pos - 1:
                if tmp == None:
                    raise Exception('Cannot insert at pos ' + str(pos))
                tmp = tmp.next
                cnt += 1
            if tmp == None:
                raise Exception('Cannot insert at pos ' + str(pos))
            new = node(data)
            new.next = tmp.next
            tmp.next = new
    
    def delete_at_beg(self):
        if self.head == None: raise Exception('List Empty ...')
        tobe = self.head
        self.head = self.head.next
        del tobe

    def delete(self, pos):
        if pos < 0: raise Exception('pos < 0')
        elif pos == 0:
            self.delete_at_beg()
        else:
            cnt = 0
            tmp = self.head
            while cnt < pos - 1:
                if tmp == None:
                    raise Exception('Cannot delete at pos ' + str(pos))
                tmp = tmp.next
                cnt += 1
            if tmp == None or tmp.next == None:
                raise Exception('Cannot delete at pos ' + str(pos))
            tobe = tmp.next
            tmp.next = tmp.next.next
            del tobe
    
    def find(self, data):
        tmp = self.head
        while tmp != None and tmp.data != data:
            tmp = tmp.next
        return tmp

    def reverse(self):
        prev = None
        nxt = None
        curr = self.head
        while curr != None:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
        self.head = prev

    def reverse_start_to_end(self, start, end):
        prev = None
        curr= start
        nxt = None
        while curr != end:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
        curr.next = prev

    def find_middle(self):
        slow_ptr = fast_ptr = self.head
        while fast_ptr != None:
            fast_ptr = fast_ptr.next
            if fast_ptr == None:
                return slow_ptr
            fast_ptr = fast_ptr.next
            if fast_ptr == None:
                return slow_ptr
            slow_ptr = slow_ptr.next

        return slow_ptr

    def kth_from_last(self, k):
        if k <= 0: return None

        slow_ptr = fast_ptr = self.head
        cnt = 0
        while cnt < k:
            if fast_ptr == None: return None
            fast_ptr = fast_ptr.next
            cnt += 1
        while fast_ptr != None:
            fast_ptr = fast_ptr.next
            slow_ptr = slow_ptr.next
        return slow_ptr

    def loop_exists(self):
        slow_ptr = fast_ptr = self.head
        while fast_ptr != None:
            fast_ptr = fast_ptr.next
            if fast_ptr == slow_ptr:
                return True
            if fast_ptr == None:
                return False
            fast_ptr = fast_ptr.next
            if fast_ptr == slow_ptr:
                return True
            if fast_ptr == None:
                return False
           
            slow_ptr = slow_ptr.next
        return None

    def reverse_blocks_of_k_nodes(self, k):
        if k <= 1: return
        tmp = node(float("-inf"))
        tmp.next = self.head
        prev = tmp
        while True:
            end = prev
            start = prev.next
            cnt = 0
            while cnt < k:
                end = end.next
                if end == None:
                    self.head = tmp.next
                    return
                cnt += 1
            nb_start = end.next
            self.reverse_start_to_end(start, end)
            start.next = nb_start
            prev.next = end
            prev = start

    def modular_node(self, k):
        if k <= 0: return None
        cnt = 1
        tmp = self.head
        modnode = None
        while tmp != None:
            if cnt % k == 0:
                modnode = tmp
            tmp = tmp.next
            cnt += 1
        return modnode

    def sqrt_node(self):
        sqrtnode = None
        cnt = 1
        tmp = self.head
        while tmp != None:
            sqrt = cnt ** 0.5
            if int(sqrt) == sqrt:
                if sqrtnode == None: sqrtnode = self.head
                else: sqrtnode = sqrtnode.next
            tmp = tmp.next
            cnt += 1
        return sqrtnode

    def fractional_node(self, k):
        if k <= 0: return None
        fnode = None
        cnt = 1
        tmp = self.head
        while tmp != None:
            if cnt % k == 0:
                if fnode == None: fnode = self.head
                else: fnode = fnode.next
            tmp = tmp.next
            cnt += 1
        return fnode

    def reorder_list(self):
        s = list()
        tmp = self.head
        while tmp != None:
            s.append(tmp)
            tmp = tmp.next
        l = head = self.head
        fromstack = True
        while (fromstack and s[-1] != l) or (not fromstack and l != head):
            if fromstack:
                head = head.next
                l.next = s.pop()
                l = l.next
                fromstack = False
            else:
                l.next = head
                l = l.next
                fromstack = True
        l.next = None

    def print_list(self):
        tmp = self.head
        while tmp != None:
            print(tmp.data, end = ' ')
            tmp = tmp.next
        print()

l = sll()
for item in range(10,14):
    l.insert(0, item)
l.print_list()
l.insert(1, 100)
l.print_list()
l.reverse()
l.print_list()
l.reverse_blocks_of_k_nodes(3)
l.print_list()
l.reorder_list()
l.print_list()
l.reorder_list()
l.print_list()
print(l.fractional_node(6))
print(l.fractional_node(5).data)
print(l.sqrt_node().data)
print(l.modular_node(1).data)
print(l.kth_from_last(0))
print(l.find_middle().data)
print(l.find(100))
l.delete(1)
l.print_list()
print(l.find_middle().data)
l.delete(0)
l.print_list()
print(l.sqrt_node().data)
print(l.find(10))
print(l.find_middle().data)
# Create a loop
n1 = node(1)
n2 = node(2)
n1.next = n2
n3 = node(3)
n2.next = n3
n4 = node(4)
n3.next = n4
n5 = node(5)
n4.next = n5
n5.next = n3
ll = sll()
ll.head = n1
print(ll.loop_exists())
print(l.loop_exists())
```

```python
def finding_spans(a):
    n = len(a)
    s = [0 for _ in range(n)]
    for i in range(n):
        s[i] = 1
        for j in range(i-1, -1, -1):
            if a[j] <= a[i]:
                s[i] += 1
            else:
                break
    return s

import random
%timeit finding_spans([i for i in range(1000)])
```

```python
def finding_spans_optimal(a):
    s = list()
    stack = list()
    n = len(a)
    for i in range(n):
        while len(stack) > 0 and a[i] >= a[stack[-1]]:
            stack.pop()
        if len(stack) == 0: j = -1
        else: j = stack[-1]
        s.append(i - j)
        stack.append(i)
    return s

import random
%timeit finding_spans([i for i in range(1000)])
```

```python
def remove_adj_duplicates(a):
    j = 0
    i = 1
    n = len(a)
    while i < n:
        if a[i] != a[j]:
            j += 1
            a[j] = a[i]
            i += 1
        else:
            del a[i]
            del a[j]
            n = n - 2
            if j > 0:
                j = j - 1
                i = i - 1
    return a

print(''.join(remove_adj_duplicates(list('careermonk'))))
print(''.join(remove_adj_duplicates(list('mississippi'))))
print(''.join(remove_adj_duplicates(list('aabb'))))
print(''.join(remove_adj_duplicates(list('g'))))
print(''.join(remove_adj_duplicates(list(''))))
```

```python
class node():
    def __init__(self, data = None, next = None, prev = None):
        self.data = data
        self.next = next
        self.prev = prev

class dll():
    def __init__(self):
        self.head = None

    def insert_at_beg(self, data):
        new = node(data)
        new.next = self.head
        if self.head != None: self.head.prev = new
        self.head = new

    def insert(self, pos, data):
        if pos < 0: raise Exception('pos < 0')
        elif pos == 0:
            self.insert_at_beg(data)
        else:
            tmp = self.head
            cnt = 0
            while cnt < pos - 1:
                if tmp == None:
                    raise Exception('Cannot insert at pos ' + str(pos))
                tmp = tmp.next
                cnt += 1
            if tmp == None:
                raise Exception('Cannot insert at pos ' + str(pos))
            new = node(data)
            new.next = tmp.next
            if tmp.next != None: tmp.next.prev = new
            tmp.next = new
            new.prev = tmp
    
    def delete_at_beg(self):
        if self.head == None: raise Exception('List Empty ...')
        tobe = self.head
        self.head = self.head.next
        if self.head != None: self.head.prev = None
        del tobe

    def delete(self, pos):
        if pos < 0: raise Exception('pos < 0')
        elif pos == 0:
            self.delete_at_beg()
        else:
            cnt = 0
            tmp = self.head
            while cnt < pos - 1:
                if tmp == None:
                    raise Exception('Cannot delete at pos ' + str(pos))
                tmp = tmp.next
                cnt += 1
            if tmp == None or tmp.next == None:
                raise Exception('Cannot delete at pos ' + str(pos))
            tobe = tmp.next
            tmp.next = tmp.next.next
            if tmp.next != None: tmp.next.prev = tmp
            del tobe

    def print_list(self):
        tmp = self.head
        while tmp != None:
            print(tmp.data, end = ' ')
            tmp = tmp.next
        print()

l = dll()
for item in range(10,14):
    l.insert(0, item)
l.print_list()
l.insert(1, 100)
l.print_list()
l.delete(1)
l.print_list()
l.delete(0)
l.print_list()
tmp = l.head.next.next
while tmp != None:
    print(tmp.data, end = ' ')
    tmp = tmp.prev
print()
```

```python
def lcsub(s, t):
    m = len(s)
    n = len(t)
    lcs = [[0 for _ in range(n+1)] for _ in range(m+1)]
    for i in range(1,m+1):
        for j in range(1, n+1):
            if s[i-1] == t[j-1]:
                lcs[i][j] = 1 + lcs[i-1][j-1]
            else:
                lcs[i][j] = max(lcs[i-1][j], lcs[i][j-1])
    match = ''
    i = m
    j = n
    while i > 0 and j > 0:
        if lcs[i][j] == lcs[i-1][j]:
            i = i - 1
        elif lcs[i][j] == lcs[i][j-1]:
            j = j - 1
        elif lcs[i][j] == 1 + lcs[i-1][j-1]:
            match = s[i-1] + match
            i = i - 1
            j = j - 1
    return (lcs[m][n], match)

s = 'ganesa'
t = 'resna'
print(lcsub(s,t))
```

```python
def lcsstring(s, t):
    m = len(s)
    n = len(t)
    lcs = [[0 for _ in range(n+1)] for _ in range(m+1)]
    maxn = float("-inf")
    maxi = float("-inf")
    maxj = float("-inf")
    for i in range(1,m+1):
        for j in range(1, n+1):
            if s[i-1] == t[j-1]:
                lcs[i][j] = 1 + lcs[i-1][j-1]
                if lcs[i][j] > maxn:
                    maxn = lcs[i][j]
                    maxi = i-1
                    maxj = j-1
            else:
                lcs[i][j] = 0

    #print(lcs)
    return [maxn, s[maxi-maxn+1:maxi+1]]

s = 'ganesa'
t = 'asz'
print(lcsstring(s,t))
```

```python
def lps(s):
    n = len(s)
    lps = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n): lps[i][i] = 1
    for i in range(n-1, -1, -1):
        for j in range(0, n):
            if i < j:
                if s[i] == s[j]:
                    if i+1 <= j-1:
                        lps[i][j] = 2 + lps[i+1][j-1]
                    else:
                        lps[i][j] = 2
                else:
                    lps[i][j] = max(lps[i+1][j], lps[i][j-1])
    match = []
    i = 0
    j = n-1
    while i <= j:
        if lps[i][j] == lps[i+1][j]:
            i = i + 1
        elif lps[i][j] == lps[i][j-1]:
            j = j - 1
        elif lps[i][j] == 2 + lps[i+1][j-1]:
            tmp = len(match) // 2
            match.insert(tmp, s[i])
            match.insert(tmp, s[i])
            i = i + 1
            j = j - 1
        elif lps[i][j] == 1 + lps[i+1][j-1]:
            tmp = len(match) // 2
            match.insert(tmp, s[i])
            i = i + 1
            j = j - 1

    return (lps[0][n-1], ''.join(match))

s = 'ganga'
print(lps(s))
print(lcsub(s, list(reversed(s))))
```

```python
def edit_distance(s, t):
    m = len(s)
    n = len(t)
    d = [[0 for _ in range(n+1)] for _ in range(m+1)]
    for i in range(m+1):
        d[i][0] = i
    for j in range(n+1):
        d[0][j] = j
    for i in range(1,m+1):
        for j in range(1,n+1):
            if s[i-1] == t[j-1]:
                d[i][j] = d[i-1][j-1]
            else:
                d[i][j] = min(d[i-1][j] + 1, d[i][j-1] + 1, d[i-1][j-1] + 2)
    return d[m][n]

s = 'ganesa'
t = 'as'
print(edit_distance(s,t))
```

```python
# Integer Knapsack
import functools
@functools.lru_cache()
def integer_knapsack(w, v, m):
    if m < 0: return float("-inf")
    elif m == 0: return 0
    maxi = float("-inf")
    for i in range(len(w)):
        maxi = max(maxi, integer_knapsack(w, v, m-w[i]) + v[i])
    maxi = max(integer_knapsack(w, v, m-1), maxi)
    return maxi

w = (1,5,10,20,25)
v = (1,20,30,40,155)
m = 40
#%timeit min_coins_exchange(w, m)
print(integer_knapsack(w, v, m))
```

```python
def gen_bin_strings(n):
    if n <= 0:
        return []
    elif n == 1:
        return ['0', '1']
    else:
        res = []
        for item in gen_bin_strings(n-1):
            for ele in ['0', '1']:
                res.append(item + ele)
        return res

print(gen_bin_strings(3))
```

```python
def perm(a, i, n, res):
    if i == n:
        res.append(''.join(a))
    else:
        for k in range(i, n):
            a[i], a[k] = a[k], a[i]
            perm(a, i+1, n, res)
            a[i], a[k] = a[k], a[i]

res = []
perm(list('abcd'), 0, len('abcd'), res)
print(res)
```

```python
def is_sorted_non_recur(a):
    for i in range(len(a) - 1):
        if a[i] > a[i+1]:
            return False
    return True

def is_sorted_recur(a):
    n = len(a)
    if n == 0 or n == 1: return True
    return a[0] <= a[1] and is_sorted_recur(a[1:])

a = [11,22,44,55,77,99]
b = [1,2,3,5,4,6,7]
c = [99,77,55,33,11]
print(is_sorted_non_recur(a))
print(is_sorted_non_recur(b))
print(is_sorted_non_recur(list(reversed(c))))
print(is_sorted_recur(a))
print(is_sorted_recur(b))
print(is_sorted_recur(list(reversed(c))))
```

```python
class content_filter():
    def __init__(self, filters = None):
        self.filters = []
        if filters is not None:
            self.filters += filters
    def filter(self, content):
        for fil in self.filters:
            content = globals()[fil]().filter(content)
        return content

class ads():
    def filter(self, content):
        print('Ads filtered.')
        return content

class offensive_words():
    def filter(self, content):
        print('Offensive Words filtered.')
        return content

class porno():
    def filter(self, content):
        print('Porno filtered.')
        return content

content = ''
cf = content_filter(['ads', 'offensive_words', 'porno'])
content = cf.filter(content)
```

```python
import os
class rename_file():
    def __init__(self, frm, to):
        self.frm = frm
        self.to = to
    def execute(self):
        os.rename(self.frm, self.to)
    def undo(self):
        os.rename(self.to, self.frm)

class history():
    def __init__(self):
        self.objs = []
    def execute(self, obj):
        self.objs.append(obj)
        obj.execute()
    def undo(self):
        obj = self.objs.pop()
        obj.undo()

h = history()
print(os.path.exists('trtr.txt'))
h.execute(rename_file('trtr.txt', 'tt.txt'))
print(os.path.exists('trtr.txt'))
print(os.path.exists('tt.txt'))
h.undo()
print(os.path.exists('trtr.txt'))
print(os.path.exists('tt.txt'))
```

```python
class subject():
    def __init__(self):
        self.sub = None
        self.observers = set()
    def attach(self, obs):
        self.observers.add(obs)
    def detach(self, obs):
        self.observers.discard(obs)
    def set_sub(self, sub):
        self.sub = sub
        self._notify()
    def _notify(self):
        for obs in self.observers:
            obs.update(self.sub)

class abc():
    def __init__(self):
        raise Exception('Cannot instantiate ABC ...')
    def update(self):
        raise Exception('Define it in the derived class ...')

class obs1(abc):
    def __init__(self):
        self.sub = None
    def update(self, sub):
        print('obs1 object value updated from {} to {}'.format(self.sub, sub))
        self.sub = sub

class obs2(abc):
    def __init__(self):
        self.sub = None
    def update(self, sub):
        print('obs2 object value updated from {} to {}'.format(self.sub, sub))
        self.sub = sub

sub = subject()
one = obs1()
two = obs2()
sub.attach(one)
sub.attach(two)
sub.set_sub(35)
sub.detach(one)
sub.set_sub(26)
```

```python
class generic():
    def __init__(self, obj, method_list):
        self.obj = obj
        self.methods = method_list
    def __getattr__(self, meth):
        if meth in self.methods:
            return getattr(self.obj, meth)
        else:
            raise Exception('Don\'t have permission to invoke method {}'.format(meth))

from queue import Queue
q = generic(Queue(), ['get', 'put', 'qsize'])
q.put(26)
print(q.get())
print(q.qsize())
s = generic(list(), ['append', 'pop'])
s.append(35)
print(s.pop())
```

```python
# Longest Common Subsequence
def longest_common_subsequence(s, t, m, n):
    lcs = [[0 for _ in range(n+1)] for _ in range(m+1)]
    for i in range(m):
        for j in range(n):
            if s[i] == t[j]:
                lcs[i+1][j+1] = 1 + lcs[i][j]
            else:
                lcs[i+1][j+1] = max(lcs[i][j+1], lcs[i+1][j])

    match = ''
    i = m
    j = n
    while i > 0 and j > 0:
        if lcs[i][j] == lcs[i-1][j]:
            i = i - 1
        elif lcs[i][j] == lcs[i][j-1]:
            j = j - 1
        elif lcs[i][j] == 1 + lcs[i-1][j-1]:
            if s[i-1] == t[j-1]:
                match = s[i-1] + match
                i = i - 1
                j = j - 1

    return (lcs[m][n], match)

s = 'ganesa'
t = 'resna'
print(longest_common_subsequence(s, t, len(s), len(t)))
```

```python
# Edit Distance
def edit_distance(s, t, m, n):
    ed = [[0 for _ in range(n+1)] for _ in range(m+1)]
    for i in range(m+1):
        ed[i][0] = i
    for j in range(n+1):
        ed[0][j] = j
    for i in range(1, m+1):
        for j in range(1, n+1):
            if s[i-1] == t[j-1]:
                ed[i][j] = ed[i-1][j-1]
            else:
                #print(i, j)
                ed[i][j] = min(ed[i-1][j] + 1, ed[i][j-1] + 1, ed[i-1][j-1] + 2)
    
    return ed[m][n]

s = 'ganesa'
t = 'asz'
print(edit_distance(s, t, len(s), len(t)))
```

```python
# Singly Linked List
class node():
    def __init__(self, data=None):
        self.data = data
        self.next = None

class singly_linked_list():
    def __init__(self):
        self.head = None
    
    def insert_at_beg(self, data):
        tmpnode = node(data)
        tmpnode.next = self.head
        self.head = tmpnode
    
    def insert(self, pos, data):
        if pos < 0:
            print('Position cannot be less than zero')
            return
        elif pos == 0:
            self.insert_at_beg(data)
        else:
            tmp = self.head
            cnt = 0
            while cnt < pos - 1:
                if tmp == None:
                    print('Cannot insert at position {}'.format(pos))
                    return
                tmp = tmp.next
                cnt += 1
            if tmp == None:
                print('Cannot insert at position {}'.format(pos))
                return
            tmpnode = node(data)
            tmpnode.next = tmp.next
            tmp.next = tmpnode
    
    def delete_at_beg(self):
        if self.head == None:
            print('List Empty. Cannot Delete')
            return
        tmp = self.head
        self.head = self.head.next
        del tmp
    
    def delete(self, pos):
        if pos < 0:
            print('Position cannot be less than zero.')
            return
        elif pos == 0:
            self.delete_at_beg()
        else:
            tmp = self.head
            cnt = 0
            while cnt < pos - 1:
                if tmp == None:
                    print('Cannot delete from position {}'.format(pos))
                    return
                tmp = tmp.next
                cnt += 1
            if tmp == None or tmp.next == None:
                print('Cannot delete from position {}'.format(pos))
                return
            delnode = tmp.next
            tmp.next = tmp.next.next
            del delnode

    def find(self, data):
        tmp = self.head
        cnt = 0
        while tmp != None and tmp.data != data:
            tmp = tmp.next
            cnt += 1
        if tmp == None: return tmp
        else: return cnt

    def remove(self, data):
        if self.head == None: return
        elif self.head.data == data:
            self.delete_at_beg()
        else:
            tmp = self.head
            while tmp.next != None and tmp.next.data != data:
                tmp = tmp.next
            if tmp.next == None: return
            else:
                remnode = tmp.next
                tmp.next = tmp.next.next
                del remnode

    def print_list(self):
        tmp = self.head
        while tmp != None:
            print(tmp.data, end = ' ')
            tmp = tmp.next
        print()

l = singly_linked_list()
for item in range(5):
    l.insert(0, item)
l.insert(5,7)
l.print_list()
print(l.find(1))
l.delete(4)
l.print_list()
l.remove(2)
l.print_list()
```

```python
# Doubly Linked List
class dnode():
    def __init__(self, data=None):
        self.data = data
        self.next = None
        self.prev = None

class doubly_linked_list():
    def __init__(self):
        self.head = None
    
    def insert_at_beg(self, data):
        tmpnode = node(data)
        tmpnode.next = self.head
        if self.head != None:
            self.head.prev = tmpnode
        self.head = tmpnode

    def insert(self, pos, data):
        if pos < 0:
            print('Position cannot be less than zero')
            return
        elif pos == 0:
            self.insert_at_beg(data)
        else:
            tmp = self.head
            cnt = 0
            while cnt < pos - 1:
                if tmp == None:
                    print('Cannot insert at position {}'.format(pos))
                    return
                tmp = tmp.next
                cnt += 1
            if tmp == None:
                print('Cannot insert at position {}'.format(pos))
                return
            tmpnode = node(data)
            tmpnode.next = tmp.next
            if tmp.next != None:
                tmp.next.prev = tmpnode
            tmp.next = tmpnode
            tmpnode.prev = tmp

    def delete_at_beg(self):
        if self.head == None:
            print('List Empty. Cannot Delete')
            return
        tmp = self.head
        self.head = self.head.next
        self.head.prev = None
        del tmp

    def delete(self, pos):
        if pos < 0:
            print('Position cannot be less than zero.')
            return
        elif pos == 0:
            self.delete_at_beg()
        else:
            tmp = self.head
            cnt = 0
            while cnt < pos - 1:
                if tmp == None:
                    print('Cannot delete from position {}'.format(pos))
                    return
                tmp = tmp.next
                cnt += 1
            if tmp == None or tmp.next == None:
                print('Cannot delete from position {}'.format(pos))
                return
            delnode = tmp.next
            tmp.next = tmp.next.next
            if tmp.next != None:
                tmp.next.prev = tmp
            del delnode

    def find(self, data):
        tmp = self.head
        cnt = 0
        while tmp != None and tmp.data != data:
            tmp = tmp.next
            cnt += 1
        if tmp == None: return tmp
        else: return cnt

    def remove(self, data):
        if self.head == None: return
        elif self.head.data == data:
            self.delete_at_beg()
        else:
            tmp = self.head
            while tmp.next != None and tmp.next.data != data:
                tmp = tmp.next
            if tmp.next == None: return
            else:
                remnode = tmp.next
                tmp.next = tmp.next.next
                if tmp.next != None:
                    tmp.next.prev = tmp
                del remnode

    def print_list(self):
        tmp = self.head
        while tmp != None:
            print(tmp.data, end = ' ')
            tmp = tmp.next
        print()

    def print_reverse_of_list(self):
        if self.head == None: return
        tmp = self.head
        while tmp.next != None:
            tmp = tmp.next
        while tmp != self.head:
            print(tmp.data, end = ' ')
            tmp = tmp.prev
        print(tmp.data)

l = doubly_linked_list()
for item in range(5):
    l.insert(0, item)
l.insert(5,7)
l.print_list()
l.print_reverse_of_list()
print(l.find(1))
l.delete(4)
l.print_list()
l.remove(2)
l.print_list()
l.print_reverse_of_list()
```

```python
# Circular Linked List
class node():
    def __init__(self, data=None):
        self.data = data
        self.next = None

class circular_linked_list():
    def __init__(self):
        self.head = None

    def insert_at_beg(self, data):
        tmpnode = node(data)
        if self.head == None:
            tmpnode.next = tmpnode
        else:
            tmp = self.head
            while tmp.next != self.head:
                tmp = tmp.next
            tmpnode.next = tmp.next
            tmp.next = tmpnode
        self.head = tmpnode

    def insert(self, pos, data):
        if pos < 0:
            print('Position cannot be less than zero')
        elif pos == 0:
            self.insert_at_beg(data)
        elif self.head == None:
            print('Cannot insert at position {}'.format(pos))
        else:
            tmp = self.head
            cnt = 0
            while cnt < pos - 1:
                tmp = tmp.next
                if tmp == self.head:
                    print('Cannot insert at position {}'.format(pos))
                    return
                cnt += 1
            if tmp == self.head and pos > 1:
                print('Cannot insert at position {}'.format(pos))
                return
            tmpnode = node(data)
            tmpnode.next = tmp.next
            tmp.next = tmpnode

    def delete_at_beg(self):
        if self.head == None:
            print('List Empty. Cannot Delete')
        elif self.head.next == self.head:
            del self.head
            self.head = None
        else:
            tmp = self.head
            while tmp.next != self.head:
                tmp = tmp.next
            delnode = self.head
            tmp.next = self.head.next
            del delnode

    def delete(self, pos):
        if pos < 0:
            print('Position cannot be less than zero.')
        elif pos == 0:
            self.delete_at_beg()
        elif self.head == None:
            print('Cannot delete from position {}'.format(pos))
        else:
            tmp = self.head
            cnt = 0
            while cnt < pos - 1:
                tmp = tmp.next
                if tmp == self.head:
                    print('Cannot delete from position {}'.format(pos))
                    return
                cnt += 1
            if (tmp == self.head or tmp.next == self.head) and (pos > 1):
                print('Cannot delete from position {}'.format(pos))
                return
            delnode = tmp.next
            tmp.next = tmp.next.next
            del delnode

    def find(self, data):
        if self.head == None: return None
        elif self.head.data == data: return 0
        else:
            tmp = self.head
            cnt = 1
            while tmp.next != self.head and tmp.next.data != data:
                tmp = tmp.next
                cnt += 1
            if tmp.next == self.head: return None
            else: return cnt

    def remove(self, data):
        if self.head == None: return
        elif self.head.data == data:
            self.delete_at_beg()
        else:
            tmp = self.head
            while tmp.next != self.head and tmp.next.data != data:
                tmp = tmp.next
            if tmp.next == self.head: return
            else:
                remnode = tmp.next
                tmp.next = tmp.next.next
                del remnode

    def print_list(self):
        if self.head == None: return
        elif self.head.next == self.head:
            print(self.head.data)
        else:
            tmp = self.head
            while tmp.next != self.head:
                print(tmp.data, end = ' ')
                tmp = tmp.next
            print(tmp.data)

l = circular_linked_list()
for item in range(5):
    l.insert(0, item)
l.insert(5,7)
l.insert(1,77)
l.print_list()
print(l.find(1))
l.delete(1)
l.delete(4)
l.print_list()
l.remove(2)
l.remove(3)
l.print_list()
```

```python
# Stack
class node():
    def __init__(self, data=None):
        self.data = data
        self.next = None

class stack():
    def __init__(self):
        self.head = None
    
    def push(self, data):
        tmpnode = node(data)
        tmpnode.next = self.head
        self.head = tmpnode
    
    def pop(self):
        if self.head == None:
            print('Stack Empty')
            return None
        else:
            tmpnode = self.head
            tmp = self.head.data
            self.head = self.head.next
            del tmpnode
            return tmp

    def top(self):
        if self.head == None:
            return None
        else:
            return self.head.data
    
    def empty(self):
        return self.head == None

s = stack()
print(s.empty())
for item in range(5):
    s.push(item)
print(s.empty())
print('TOP')
print(s.top())
print('POP')
while not s.empty():
    print(s.pop())
print(s.empty())
```

```python
# Queue
class node():
    def __init__(self, data=None):
        self.data = data
        self.next = None

class queue():
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, data):
        tmpnode = node(data)
        if self.front == None:
            self.front = self.rear = tmpnode
        else:
            self.rear.next = tmpnode
            self.rear = tmpnode

    def dequeue(self):
        if self.front == None:
            print('Queue Empty')
            return None
        else:
            delnode = self.front
            tmp = delnode.data
            self.front = self.front.next
            if self.front == None:
                self.rear = None
            del delnode
            return tmp

    def empty(self):
        return self.front == None

q = queue()
print(q.empty())
for item in range(5):
    q.enqueue(item)
print(q.empty())
print('DEQUEUE')
while not q.empty():
    print(q.dequeue())
print(q.empty())
```

```python
# Binary Search
def binary_search_helper(a, data, low, high):
    if low <= high:
        mid = (low + high) // 2
        if data == a[mid]:
            return mid
        elif data < a[mid]:
            return binary_search_helper(a, data, low, mid - 1)
        elif data > a[mid]:
            return binary_search_helper(a, data, mid + 1, high)
    return None

def binary_search(a, data):
    return binary_search_helper(a, data, 0, len(a) - 1)

a = [11, 22, 33, 44, 55, 66, 77, 88, 99]
print(binary_search(a, 44))
print(binary_search(a, 11))
print(binary_search(a, 99))
print(binary_search(a, -999))
print(binary_search(a, 999))
```

```python
# First Occurrence of a Number
def binary_search_first_helper(a, data, low, high):
    if low <= high:
        mid = (low + high) // 2
        if data == a[mid] and (mid == low or a[mid-1] < a[mid]):
            return mid
        elif data == a[mid]:
            return binary_search_first_helper(a, data, low, mid - 1)
        elif data < a[mid]:
            return binary_search_first_helper(a, data, low, mid - 1)
        elif data > a[mid]:
            return binary_search_first_helper(a, data, mid + 1, high)
    return None

def binary_search_first(a, data):
    return binary_search_first_helper(a, data, 0, len(a) - 1)

a = [11, 11, 22, 33, 44, 44, 44, 55, 66, 77, 88, 99, 99]
print(binary_search_first(a, 44))
print(binary_search_first(a, 11))
print(binary_search_first(a, 99))
print(binary_search_first(a, -666))
print(binary_search_first(a, 666))
```

```python
# Last Occurrence of a Number
def binary_search_last_helper(a, data, low, high):
    if low <= high:
        mid = (low + high) // 2
        if data == a[mid] and (mid == high or a[mid+1] > a[mid]):
            return mid
        elif data == a[mid]:
            return binary_search_last_helper(a, data, mid + 1, high)
        elif data < a[mid]:
            return binary_search_last_helper(a, data, low, mid - 1)
        elif data > a[mid]:
            return binary_search_last_helper(a, data, mid + 1, high)
    return None

def binary_search_last(a, data):
    return binary_search_last_helper(a, data, 0, len(a) - 1)

a = [11, 11, 22, 33, 44, 44, 44, 55, 66, 77, 88, 99, 99]
print(binary_search_last(a, 44))
print(binary_search_last(a, 11))
print(binary_search_last(a, 99))
print(binary_search_last(a, -777))
print(binary_search_last(a, 777))
print(binary_search_last(a, 16))
```

```python
# Linear Search
def linear_search(a, data):
    for i in range(len(a)):
        if a[i] == data:
            return i
    return None

a = [33, 11, -100, 44, 99, 66, 100, 22, 77]
print(linear_search(a, 11))
print(linear_search(a, 77))
print(linear_search(a, 33))
print(linear_search(a, -111))
print(linear_search(a, 47))
print(linear_search(a, 111))
```

```python
# Bubble Sort
def bubble_sort(a):
    n = len(a)
    for i in range(n):
        for j in range(0, n-i-1):
            if a[j] > a[j+1]:
                a[j],a[j+1] = a[j+1],a[j]
        print(a)

def improved_bubble_sort(a):
    n = len(a)
    for i in range(n):
        toggle = 0
        for j in range(0, n-i-1):
            if a[j] > a[j+1]:
                toggle = 1
                a[j], a[j+1] = a[j+1], a[j]
        if toggle == 0: break
        print(a)

a = [33, 11, -100, 44, 99, 66, 100, 22, 77]
a = [-100, 11, 22, 33, 44, 66, 77, 99, 100]
bubble_sort(a)
print(a)
print("Improved Bubble Sort")
improved_bubble_sort(a)
print(a)
```

```python
def selection_sort(a):
    n = len(a)
    for i in range(n):
        small = i
        for j in range(i+1,n):
            if a[j] < a[small]:
                small = j
        a[small],a[i] = a[i], a[small]

a = [33, 11, -100, 44, 99, 66, 100, 22, 77]
selection_sort(a)
print(a)
```

```python
def insertion_sort(a):
    n = len(a)
    for i in range(n):
        j = i
        tmp = a[j]
        while j > 0 and a[j-1] > tmp:
            a[j] = a[j-1]
            j -= 1
        a[j] = tmp

a = [33, 11, -100, 44, 99, 66, 100, 22, 77]
insertion_sort(a)
print(a)
```

```python
def shell_sort(a):
    n = len(a)
    h = n // 2
    while h > 0:
        for i in range(n):
            j = i
            tmp = a[j]
            while j >= h and a[j-h] > tmp:
                a[j] = a[j-h]
                j -= h
            a[j] = tmp
        h = h // 2

a = [33, 11, -100, 44, 99, 66, 100, 22, 77]
shell_sort(a)
print(a)
```

```python
def merge_sort_helper(a, low, high):
    if low < high:
        mid = (low + high) // 2
        merge_sort_helper(a, low, mid)
        merge_sort_helper(a, mid + 1, high)
        tmp = [None for _ in range(high - low + 1)]
        k = 0
        i = low
        j = mid + 1
        while i <= mid and j <= high:
            if a[i] < a[j]:
                tmp[k] = a[i]
                i += 1
                k += 1
            else:
                tmp[k] = a[j]
                j += 1
                k += 1
        while i <= mid:
            tmp[k] = a[i]
            i += 1
            k += 1
        while j <= high:
            tmp[k] = a[j]
            j += 1
            k += 1
        k = k - 1
        j = high
        while k >= 0:
            a[j] = tmp[k]
            k = k - 1
            j = j - 1

def merge_sort(a):
    merge_sort_helper(a, 0, len(a) - 1)

a = [33, 11, -100, 44, 99, 66, 100, 22, 77]
merge_sort(a)
print(a)
```

```python
def get_median3(l):
    l.sort()
    return l[1]

def quick_sort_helper(a, low, high):
    if low < high:
        pivot, pivot_index = get_median3([[a[low], low], [a[high], high], [a[(low + high) // 2], (low + high) // 2]])
        a[pivot_index], a[high] = a[high], a[pivot_index]
        i = low
        j = high
        while i < j:
            while a[i] <= pivot and i < high: i += 1
            while a[j] >= pivot and j > low: j -= 1
            if i < j: a[i], a[j] = a[j], a[i]
        a[i], a[high] = a[high], a[i]
        quick_sort_helper(a, low, i - 1)
        quick_sort_helper(a, i + 1, high)

def quick_sort(a):
    quick_sort_helper(a, 0, len(a) - 1)

a = [33, 11, -100, 44, 99, 66, 100, 22, 77]
quick_sort(a)
print(a)
```

```python
# Exact substring match - Brute Force
def exact_substring_match(s, t):
    n = len(s)
    m = len(t)
    for i in range(0, n-m+1):
        for j in range(0, m):
            if s[i+j] != t[j]:
                break
        else:
            return i
    return None

s = 'bacbabababacaca'
t = 'ababaca'
print(exact_substring_match(s,t))
```

```python
# KMP Algorithm
def prefix_table(t):
    m = len(t)
    f = [0 for _ in range(m)]
    f[0] = 0
    j = 0
    for i in range(1,m):
        while j > 0 and t[i] != t[j]:
            j = f[j-1]
        if t[i] == t[j]:
            j += 1
        f[i] = j
    return f

def kmp(s, t):
    n = len(s)
    m = len(t)
    f = prefix_table(t)
    j = 0
    for i in range(0, n):
        while j > 0 and s[i] != t[j]:
            j = f[j-1]
        if s[i] == t[j]:
            if j == m - 1: return i-m+1
            j += 1
    return None

s = 'bacbabzzababacaca'
t = 'ababaca'
print(prefix_table(t))
print(kmp(s,t))
```

```python
# Rabin-Karp Algorithm
def get_hash(t):
    hasht = 0
    for i in range(len(t)):
        hasht = hasht * 75 + (ord(t[i]) - ord('0'))
    return hasht

def rka(s, t):
    n = len(s)
    m = len(t)
    hasht = get_hash(t)
    #print(hasht)
    hashs = get_hash(s[0:m])
    #print(hashs)
    if hashs == hasht:
        if s[0:m] == t:
            return 0
    for i in range(1,n-m+1):
        hashs = (hashs % (75 ** (m-1))) * 75 + (ord(s[i+m-1]) - ord('0'))
        #print(hashs)
        if hashs == hasht:
            if s[i:i+m] == t:
                return i
    return None

s = 'bacbabababacaca'
t = 'ababacaca'
s = '102345'
t = '02345'
s = 'ganesa'
t = 'esa'
print(rka(s,t))
```

```python
# Selecting kth element using 
# Median of Medians Algorithm
# Linear Time
CHUNK_SIZE = 5
def get_chunks(a):
    chunks = []
    for i in range(0,len(a),CHUNK_SIZE):
        chunks.append(a[i:i+CHUNK_SIZE].copy())
    return chunks

def kth_ele(a, k):
    n = len(a)
    if k > n or k <= 0: return None
    if n <= CHUNK_SIZE:
        a.sort()
        return a[k-1]
    chunks = get_chunks(a)
    #print(chunks)
    meds = []
    for chunk in chunks:
        chunk.sort()
        meds.append(chunk[len(chunk) // 2])
    #print(meds)
    med_of_meds = kth_ele(meds, len(meds) // 2)
    #print(med_of_meds)
    left = []
    right = []
    middle = []
    for item in a:
        if item == med_of_meds: middle.append(item)
        elif item < med_of_meds: left.append(item)
        else: right.append(item)
    #print(left, middle, right)
    if (k > len(left)) and (k <= len(left) + len(middle)): return med_of_meds
    elif k <= len(left): return kth_ele(left, k)
    else: return kth_ele(right, k - (len(left) + len(middle)))

a = [33, 11, -100, 44, 99, 66, 100, 22, 77, -234, 234, 56, 62, 844, 857]
print(sorted(a))
print(kth_ele(a, 14))
```

```python
# Kth Smallest Element given 2 sorted arrays
# log(m+n)
def kth_ele_2_sorted_arrays(a, b, k):
    m = len(a)
    n = len(b)
    if k <= 0 or k > (m + n): return None
    if m > n:
        a,b = b,a
        m,n = n,m
    stepsa = (min(k, m) - 1) // 2
    stepsb = k - stepsa - 2
    if stepsb > (n-1): stepsb = n-1
    if m == 0: return b[k-1]
    elif k == 1: return min(a[0], b[0])
    elif a[stepsa] == b[stepsb]: return a[stepsa]
    elif a[stepsa] < b[stepsb]: return kth_ele_2_sorted_arrays(a[stepsa+1:], b, k-stepsa-1)
    else: return kth_ele_2_sorted_arrays(a, b[stepsb+1:], k-stepsb-1)

a = [-234, -100, 11, 22, 33, 55, 56]
b = [-33, 6, 13, 17, 44, 47]
c = a + b
c.sort()
print(c)
print(kth_ele_2_sorted_arrays(a, b, 13))
```

```python
# Median of 2 sorted arrays (arrays are of equal length)
# log(n)
import math
def median_of_2_sorted_arrays(a, b):
    n = len(a)
    if n == 0: return None
    elif n == 1: return (a[0] + b[0]) / 2
    elif n == 2: return (max(a[0], b[0]) + min(a[1], b[1])) / 2
    else:
        meda = math.floor((n-1)/2)
        medb = math.ceil((n-1)/2)
        if a[meda] == b[medb]: return a[meda]
        elif a[meda] < b[medb]: return median_of_2_sorted_arrays(a[meda:], b[:medb+1])
        else: return median_of_2_sorted_arrays(a[:meda+1], b[medb:])

a = [-234, 11, 22, 33, 55, 56]
b = [-33, 6, 13, 44, 47, 102]
c = a + b
c.sort()
print(c)
print(median_of_2_sorted_arrays(a, b))
```

```python
# Hash Class
HTABLE_SIZE = 7
LOAD_FACTOR = 0.7
class hashe():
    def __init__(self):
        self.size = HTABLE_SIZE
        self.table = [None for _ in range(self.size)]
        self.cnt = 0
    
    def get_hash_val(self, key):
        import hashlib
        h = hashlib.sha256(key.encode('ascii'))
        return int(h.hexdigest(), base=16) % self.size

    def next_probe(self, hash_val):
        return (hash_val + 1) % self.size

    def is_prime(self, n):
        if n <= 0:
            print('Not dealing with numbers less than or equal to zero.')
            return None

        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return 0
        return 1

    def next_prime(self):
        n = self.size + 1
        while True:
            if self.is_prime(n): return n
            n += 1

    def rehash(self):
        print('Rehashing the Hash Table')
        oldsize = self.size
        oldtable = self.table.copy()
        self.size = self.next_prime()
        self.table = [None for _ in range(self.size)]
        for item in oldtable:
            if item != None:
                hash_val = self._search(item[0])
                self.table[hash_val] = item
        del oldtable

    def _search(self, key):
        key = str(key)
        hash_val = self.get_hash_val(key)
        while self.table[hash_val] != None and self.table[hash_val][0] != key:
            hash_val = self.next_probe(hash_val)
        return hash_val

    def __setitem__(self, key, value):
        key = str(key)
        hash_val = self._search(key)
        if self.table[hash_val] == None:
            self.cnt += 1
        self.table[hash_val] = [key, value]
        print(self.cnt, self.table)
        if (self.cnt / self.size) > LOAD_FACTOR:
            self.rehash()

    def __getitem__(self, key):
        key = str(key)
        hash_val = self._search(key)
        if self.table[hash_val] == None: return None
        else: return self.table[hash_val][1]

    def __delitem__(self, key):
        key = str(key)
        hash_val = self._search(key)
        if self.table[hash_val] !=  None:
            self.table[hash_val] = None
            self.cnt -= 1

h = hashe()
for item in range(10):
    h[item] = item
for item in range(10):
    h[item] = item
for item in range(10):
    print(h[item], end = ' ')
print()
print(h[13])
del h[4]
del h[2]
print(h.cnt, h.table)
```

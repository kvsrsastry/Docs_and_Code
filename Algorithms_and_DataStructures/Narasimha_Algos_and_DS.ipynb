```python
# Generating Binary Strings
def gen_binary_strings(n):
    if n == 0:
        return []
    elif n == 1:
        return ['0', '1']
    else:
        lis = []
        for item in gen_binary_strings(n-1):
            lis.append('0' + item)
            lis.append('1' + item)
        return lis

out = gen_binary_strings(3)
print(out)
```

```python
# Generating k-ary Strings
def gen_kary_strings(n, k):
    if n == 0:
        return []
    elif n == 1:
        return [ele for ele in k]
    else:
        lis = []
        for item in gen_kary_strings(n-1, k):
            for ele in k:
                lis.append(ele + item)
        return lis

out = gen_kary_strings(2, ['0','1','2'])
print(out)
```

```python
# Permutations of a String
def perm(a, i, n, res):
    if i == n:
        #res.append(a.copy())  # copy() is very important here
        res.append(''.join(a))
    else:
        for k in range(i, n):
            a[i], a[k] = a[k], a[i]
            perm(a, i+1, n, res)
            a[i], a[k] = a[k], a[i]

res = []
inp = list('abcd')
perm(inp, 0, len(inp), res)
print(len(res))
print(res)
```

```python
# Finding whether a given array is ascending order sorted
def is_sorted_non_recur(lis):
    for i in range(0, len(lis) - 1):
        if not (lis[i] <= lis[i+1]):
            return False
    return True

def is_sorted_recur(lis):
    if len(lis) <= 1:
        return True
    return lis[0] <= lis[1] and is_sorted_recur(lis[1:])

a = [11,22,44,55,77,99]
b = [1,2,3,5,4,6,7]
c = [99,77,55,33,11]
print(is_sorted_non_recur(a))
print(is_sorted_non_recur(b))
print(is_sorted_non_recur(list(reversed(c))))
print(is_sorted_recur(a))
print(is_sorted_recur(b))
print(is_sorted_recur(list(reversed(c))))
```

```python
# Singly Linked List
class node():
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

class ListException(Exception):
    def __init__(self):
        print('List Exception .. Take a look')

class sll():
    def __init__(self, head=None):
        self.head = head
    def insert_at_beg(self, data):
        tmp = node(data)
        tmp.next = self.head
        self.head = tmp
    def insert(self, pos, data):
        if pos < 0:
            raise ListException()
        if pos == 0:
            self.insert_at_beg(data)
        else:
            tmp = self.head
            cnt = 0
            while(cnt < pos - 1):
                if tmp != None:
                    tmp = tmp.next
                else:
                    raise ListException()
                cnt += 1
            if tmp == None: raise ListException()
            tmpnode = node(data)
            tmpnode.next = tmp.next
            tmp.next = tmpnode
    def print_list(self):
        tmp = self.head
        while(tmp != None):
            print(tmp.data, end = ' ')
            tmp = tmp.next
        print()
    def find(self, data):
        tmp = self.head
        while(tmp != None and tmp.data != data):
            tmp = tmp.next
        return tmp
    def delete_at_beg(self):
        if self.head == None: raise ListException
        self.head = self.head.next
    def delete(self, pos):
        if pos < 0:
            raise ListException()
        if pos == 0:
            self.delete_at_beg()
        else:
            cnt = 0
            tmp = self.head
            while(cnt < pos - 1):
                if tmp != None:
                    tmp = tmp.next
                else:
                    raise ListException()
                cnt += 1
            if tmp == None or tmp.next == None: raise ListException()
            tmp.next = tmp.next.next

    def reverse(self):
        if self.head == None:
            return
        prev = None
        while(self.head != None):
            nxt = self.head.next
            self.head.next = prev
            prev = self.head
            self.head = nxt

        self.head = prev
    
    def find_middle(self):
        slowptr = fastptr = self.head
        while(fastptr != None):
            fastptr = fastptr.next
            if(fastptr == None):
                break
            fastptr = fastptr.next
            if(fastptr == None):
                break
            slowptr = slowptr.next

        return slowptr
    
    def find_kth_from_end(self, k):
        slowptr = fastptr = self.head
        cnt = 0
        while(fastptr != None and cnt < k):
            fastptr = fastptr.next
            cnt += 1
        if fastptr == None and cnt < k: return None
        while(fastptr != None):
            fastptr = fastptr.next
            slowptr = slowptr.next
        return slowptr
    
    def loop_exists(self):
        slowptr = fastptr = self.head
        while(fastptr != None):
            fastptr = fastptr.next
            if fastptr == slowptr:
                return 1
            if fastptr == None:
                return 0
            fastptr = fastptr.next
            if fastptr == slowptr:
                return 1
            if fastptr == None:
                return 0
            slowptr = slowptr.next
        return 0
    
    def reverse_list_1(self, start, end):
        prev = None
        cur = start
        while cur != end:
            nxt = cur.next
            cur.next = prev
            prev = cur
            cur = nxt
        cur.next = prev

    def reverse_blocks_of_k_nodes(self, k):
        temp = node()
        temp.next = self.head
        prev = temp
        while True:
            begin = prev.next
            end = prev
            cnt = 0
            while(cnt < k):
                end = end.next
                if end == None:
                    self.head = temp.next
                    return
                cnt += 1
            nextblock = end.next
            self.reverse_list_1(begin, end)
            prev.next = end
            begin.next = nextblock
            prev = begin
            
    def modular_node(self, k):
        if k <= 0: return None
        i = 1
        tmp = self.head
        modnode = None
        while(tmp != None):
            if i % k == 0:
                modnode = tmp
            tmp = tmp.next
            i += 1
        if modnode == None: return None
        else: return modnode.data

    def sqrt_nth_node(self):
        i = 1
        tmp = self.head
        dnode = None
        while(tmp != None):
            temp = i ** 0.5
            if temp == int(temp):
                if dnode == None:
                    dnode = self.head
                else:
                    dnode = dnode.next
            tmp = tmp.next
            i += 1
        if dnode == None: return None
        else: return dnode.data
    
    def fractional_node(self, k):
        if k <= 0: return None
        tmp = self.head
        dnode = None
        i = 1
        while tmp!= None:
            if i%k == 0:
                if dnode == None:
                    dnode = self.head
                else:
                    dnode = dnode.next
            tmp = tmp.next
            i += 1

        if dnode == None and (i-1) < k: return self.head.data
        if dnode == None: return None
        else: return dnode.data

    def reorder_list(self):
        head = self.head
        stack = []
        while(head != None):
            stack.append(head)
            head = head.next
        fromstack = True
        fromhead = self.head
        newlist = self.head
        while((fromstack and  newlist != stack[-1]) or (not fromstack and newlist != fromhead)):
            if fromstack:
                fromhead = fromhead.next
                newlist.next = stack.pop()
                fromstack = False
            else:
                newlist.next = fromhead
                fromstack = True
            newlist = newlist.next
        newlist.next = None

l = sll()
l.insert(0,12)
l.insert(1,13)
l.insert(2,22)
l.print_list()
l.insert(2,55)
l.insert(0,11)
l.print_list()
l.reorder_list()
l.print_list()
print(l.modular_node(5))
print(l.sqrt_nth_node())
print(l.fractional_node(7))
print(l.find_middle().data)
print(l.find_kth_from_end(2).data)
l.reverse_blocks_of_k_nodes(4)
l.print_list()
l.reverse_blocks_of_k_nodes(2)
l.print_list()
l.reverse()
l.print_list()
print(l.find(-1))
print(l.find(55))
l.delete(2)
l.print_list()
print(l.find_middle().data)
l.delete(3)
l.print_list()
print(l.find_middle().data)
l.delete(0)
l.print_list()
l.reverse()
l.print_list()
print(l.find_middle().data)
print(l.sqrt_nth_node())
print(l.fractional_node(4))
# Create a loop
n1 = node(1)
n2 = node(2)
n1.next = n2
n3 = node(3)
n2.next = n3
n4 = node(4)
n3.next = n4
n5 = node(5)
n4.next = n5
n5.next = n3
ll = sll()
ll.head = n1
print(ll.loop_exists())
print(l.loop_exists())
```

```python
# Finding Spans
def find_spans(a):
    out = []
    for i in range(len(a)):
        span = 1
        for j in range(i):
            if a[j] <= a[i]:
                span += 1
        out.append(span)
    return out

print(find_spans(list(range(1000))))
```

```python
# Finding Spans
def find_spans_stack(a):
    stack = []
    out = []
    for i in range(0, len(a)):
        while(len(stack) > 0 and a[i] >= a[stack[-1]]):
            stack.pop()
        if len(stack) == 0:
            j = -1
        else:
            j = stack[-1]
        out.append(i - j)
        stack.append(i)
    return out

print(find_spans_stack(list(range(1000))))
```

```python
def remove_adj_duplicates(s):
    if len(s) <= 1: return s
    i = 1
    j = 0
    while(i < len(s)):
        if(s[i] != s[j]):
            j += 1
            s[j] = s[i]
            i += 1
        else:
            del s[i]
            del s[j]
            if j > 0:
                j = j-1
                i = i-1
    
    return s

print(''.join(remove_adj_duplicates(list('careermonk'))))
print(''.join(remove_adj_duplicates(list('mississippi'))))
print(''.join(remove_adj_duplicates(list('aabb'))))
```

```python
# Priority Queues (Heaps)
class HeapException(Exception):
    pass

class maxheap():
    def __init__(self):
        self.arr = []
        self.size = 0
    def left_child(self, i):
        lchild = 2*i + 1
        if lchild < self.size:
            return lchild
        else:
            return float("-inf")
    def right_child(self, i):
        rchild = 2*i + 2
        if rchild < self.size:
            return rchild
        else:
            return float("-inf")
    def parent(self, i):
        if i <= 0 or i > self.size - 1:
            raise HeapException
        else:
            return (i-1)//2
    def perc_down(self, i):
        if i < 0 or i > self.size - 1: raise HeapException
        ele = self.arr[i]
        maxa = None
        maxi = None
        while(True):
            lchild = self.left_child(i)
            rchild = self.right_child(i)
            if lchild == float("-inf") and rchild == float("-inf"):
                break
            elif rchild == float("-inf"):
                maxa = self.arr[lchild]
                maxi = lchild
                if ele > maxa:
                    break
                else:
                    self.arr[i] = maxa
                    i = maxi
            else:
                if self.arr[lchild] > self.arr[rchild]:
                    maxa = self.arr[lchild]
                    maxi = lchild
                else:
                    maxa = self.arr[rchild]
                    maxi = rchild
                if ele > maxa:
                    break
                else:
                    self.arr[i] = maxa
                    i = maxi
        self.arr[i] = ele

    def heapify(self, lis):
        self.arr = lis.copy()
        self.size = len(self.arr)
        for i in range((self.size//2) - 1, -1, -1):
            self.perc_down(i)
    
    def print_heap(self):
        print(self.arr)

    def delete_max(self):
        if self.size == 0:
            raise HeapException
        tmp = self.arr[0]
        self.arr[0] = self.arr[self.size - 1]
        del self.arr[self.size - 1]
        self.size -= 1
        if self.size > 0: self.perc_down(0)
        return tmp

    def get_max(self):
        if self.size == 0:
            raise HeapException
        return self.arr[0]

    def perc_up(self, i):
        ele = self.arr[i]
        while(i > 0):
            par = self.parent(i)
            if ele > self.arr[par]:
                self.arr[i] = self.arr[par]
                i = par
            else:
                break
        self.arr[i] = ele

    def insert(self, data):
        self.arr.append(data)
        self.size += 1
        self.perc_up(self.size - 1)

array = [1,4,3,2,17,6,15]
mheap = maxheap()
mheap.heapify(array)
mheap.print_heap()
print(mheap.delete_max())
mheap.print_heap()
print(mheap.delete_max())
mheap.print_heap()
mheap1 = maxheap()
for i in array:
    mheap1.insert(i)
mheap1.print_heap()
while mheap1.size > 0:
    print(mheap1.delete_max(), end = ' ')
print()
```

```python
# Given 2 sorted lists on size 'n', find median of the '2n' elements in O(logn)
import math
def median(a,b):
    tot_size = len(a) + len(b)
    lowa = 0
    higha = len(a) - 1
    lowb = 0
    highb = len(b) - 1
    #print(a,b,lowa,higha,lowb,highb)
    #print(tot_size)
    if tot_size == 0:
        return None
    elif tot_size == 2:
        return (a[0] + b[0]) / 2
    elif tot_size == 4:
        return (max(a[0], b[0]) + min(a[1], b[1])) / 2
    else:
        meda = math.ceil((higha - lowa) / 2)    # math.ceil is the key in case of even sized block
        medb = math.floor((highb - lowb) / 2)   # math.floor is the key in case of even sized block
        #print("meda = {}".format(meda))
        #print("medb = {}".format(medb))
        if a[meda] == b[medb]:
            return a[meda]
        elif a[meda] < b[medb]:
            return median(a[meda:higha+1],b[lowb:medb+1])
        else:
            return median(a[lowa:meda+1],b[medb:highb+1])

a = [10,20,40,90,100,120,140]
b = [30,50,70,110,200,300,400]
print(median(a,b))
```

```python
# Combinations of a String
def comb(a):
    if len(a) == 0:
        return []
    elif len(a) == 1:
        return [a]
    else:
        out = []
        tmp = comb(a[1:])
        out.extend(tmp)
        for item in tmp:
            out.append(a[0:1] + item)
        out.append(a[0:1])
        return out

print(comb("abc"))
```

```python
# Permutations of a String
def perm(a, i, n, res):
    if i == n:
        res.append(''.join(a.copy()))
    else:
        for k in range(i, n):
            a[i], a[k] = a[k], a[i]
            perm(a, i+1, n, res)
            a[i], a[k] = a[k], a[i]
            
a = "abc"
res = []
perm(list(a), 0, len(a), res)
print(len(res))
print(res)
```

```python
# Insertion Sort
def insertion_sort(a):
    for i in range(1, len(a)):
        tmp = a[i]
        k = i
        while(k > 0 and a[k-1] > tmp):
            a[k] = a[k-1]
            k = k - 1
        a[k] = tmp

a = [99,22,44,11,33,77,55,66]
insertion_sort(a)
print(a)
```

```python
# Selection Sort
def selection_sort(a):
    for i in range(len(a)):
        mini = i
        for j in range(i+1, len(a)):
            if a[j] < a[mini]:
                mini = j
        a[i], a[mini] = a[mini], a[i]

a = [99,22,44,11,33,77,55,66]
selection_sort(a)
print(a)
```

```python
# Bubble Sort
def bubble_sort(a):
    for i in range(len(a)):
        for k in range(len(a) - 1, i, -1):
            if a[k] < a[k-1]:
                a[k], a[k-1] = a[k-1], a[k]

a = [99,22,44,11,33,77,55,66]
bubble_sort(a)
print(a)
```

```python
# Shell Sort
def shell_sort(a):
    h = len(a) // 2
    while(h > 0):
        for i in range(h, len(a)):
            tmp = a[i]
            k = i
            while(k >= h and a[k-h] > tmp):
                a[k] = a[k-h]
                k = k - h
            a[k] = tmp
        h = h // 2

a = [99,22,44,11,33]
shell_sort(a)
print(a)
```

```python
# Binary Tree
class BinaryTreeNode():
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None

#Use Level Order Traversal
def insert(t, data):
    if t == None:
        tmp = BinaryTreeNode(data)
        return tmp
    from queue import Queue
    q = Queue()
    q.put(t)
    tnode = BinaryTreeNode(data)
    while(not q.empty()):
        tmp = q.get()
        if tmp.left and tmp.right:
            q.put(t.left)
            q.put(t.right)
        elif tmp.left == None:
            tmp.left = tnode
            break
        elif tmp.right == None:
            tmp.right = tnode
            break
    return t

def preorder(t):
    if(t):
        print(t.data, end = ' ')
        preorder(t.left)
        preorder(t.right)

def inorder(t):
    if t:
        inorder(t.left)
        print(t.data, end = ' ')
        inorder(t.right)

def postorder(t):
    if t:
        postorder(t.left)
        postorder(t.right)
        print(t.data, end = ' ')

def levelorder(t):
    from queue import Queue
    if t == None:
        return
    q = Queue()
    q.put(t)
    while(not q.empty()):
        tmp = q.get()
        print(tmp.data, end = ' ')
        if tmp.left != None:
            q.put(tmp.left)
        if tmp.right != None:
            q.put(tmp.right)

def find(t, data):
    if t == None:
        return 0
    elif t.data == data:
        return 1
    else:
        return find(t.left, data) or find(t.right, data)

def maxi(t):
    if t == None:
        return float("-inf")
    max_left = maxi(t.left)
    max_right = maxi(t.right)
    return max([t.data, max_left, max_right])

def size(t):
    if t == None:
        return 0
    return (1 + size(t.left) + size(t.right))

def level_order_reverse(t):
    if t == None:
        return
    from queue import Queue
    q = Queue()
    stack = []
    q.put(t)
    while(not q.empty()):
        tmp = q.get()
        stack.append(tmp.data)
        if tmp.right != None:
            q.put(tmp.right)
        if tmp.left != None:
            q.put(tmp.left)
    while(len(stack) != 0):
        print(stack.pop(), end = ' ')
    print()

def delete_tree(t):
    if t == None:
        return t
    else:
        t.left = delete_tree(t.left)
        t.right = delete_tree(t.right)
        del t
        return None

def height(t):
    if t == None:
        return -1
    else:
        return(1 + max(height(t.left), height(t.right)))

def find_a_leaf(t):
    if t == None:
        return 0
    if t.left == None and t.right == None:
        return t
    left = find_a_leaf(t.left)
    if(left != 0): return left
    return find_a_leaf(t.right)

def delete(t, data):
    if t == None:
        return None
    elif t.data == data:
        leaf = find_a_leaf(t)
        if leaf == t:
            del t
            return None
        else:
            tmp = leaf.data
            t.left = delete(t.left, tmp)
            t.right = delete(t.right, tmp)
            t.data = tmp
            return t
    else:
        t.left = delete(t.left, data)
        t.right = delete(t.right, data)
        return t

def num_leaves(t):
    if t == None:
        return 0
    elif t.left == None and t.right == None:
        return 1
    else:
        return num_leaves(t.left) + num_leaves(t.right)

def full_nodes(t):
    if t == None:
        return 0
    elif t.left == None:
        return full_nodes(t.right)
    elif t.right == None:
        return full_nodes(t.left)
    else:
        return 1 + full_nodes(t.left) + full_nodes(t.right)

def half_nodes(t):
    if t == None:
        return 0
    elif t.left == None and t.right == None:
        return 0
    elif t.left and t.right:
        return half_nodes(t.left) + half_nodes(t.right)
    elif t.left == None and t.right != None:
        return 1 + half_nodes(t.right)
    elif t.right == None and t.left != None:
        return 1 + half_nodes(t.left)

class Tree():
    def __init__(self, root=None):
        self.root = root

    def insert(self, data):
        self.root = insert(self.root, data)
    
    def preorder(self):
        preorder(self.root)
        
    def inorder(self):
        inorder(self.root)
    
    def postorder(self):
        postorder(self.root)

    def levelorder(self):
        levelorder(self.root)
        
    def find(self, data):
        return find(self.root, data)
    
    def delete_tree(self):
        self.root = delete_tree(self.root)

    def delete(self, data):
        self.root = delete(self.root, data)

    def lca_helper(self, t, data1, data2):
        if t == None:
            return None
        elif t.data == data1 or t.data == data2:
            return t
        else:
            left = self.lca_helper(t.left, data1, data2)
            right = self.lca_helper(t.right, data1, data2)
            if left and right:
                return t
            else:
                return left if left else right

    def lca(self, data1, data2):
        return self.lca_helper(self.root, data1, data2)
    
t = Tree()
print(type(t))
t.insert(77)
t.insert(33)
t.insert(44)
t.insert(55)
t.insert(99)
t.insert(11)
t.preorder()
print()
t.inorder()
print()
t.postorder()
print()
t.levelorder()
print()
#t.delete(44)
#t.delete(33)
#t.delete(55)
#t.delete(11)
#t.delete(99)
t.preorder()
print()
t.inorder()
print()
t.postorder()
print()
t.levelorder()
print()
print('LCA ' + str(t.lca(11,33).data))
print(t.find(33))
print(t.find(100))
print(maxi(t.root))
print(size(t.root))
level_order_reverse(t.root)
#t.delete_tree()
print(height(t.root))
print(num_leaves(t.root))
print(full_nodes(t.root))
print(half_nodes(t.root))
```

```python
# AVL Tree - HB(1)
class AVLNode():
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None
        self.height = 0

class AVLTree():
    def __init__(self):
        self.root = None

    def height(self, t):
        if t == None:
            return -1
        else:
            return t.height

    def single_rotate_left_to_right(self, t):
        d = t.left
        t.left = d.right
        d.right = t
        t.height = max(self.height(t.left), self.height(t.right)) + 1
        return d
    
    def single_rotate_right_to_left(self, t):
        w = t.right
        t.right = w.left
        w.left = t
        t.height = max(self.height(t.left), self.height(t.right)) + 1
        return w

    def double_rotate_right_left(self, t):
        t.left = self.single_rotate_right_to_left(t.left)
        return self.single_rotate_left_to_right(t)
    
    def double_rotate_left_right(self, t):
        t.right = self.single_rotate_left_to_right(t.right)
        return self.single_rotate_right_to_left(t)
    
    def inserth(self, t, data):
        if t == None:
            t = AVLNode(data)

        elif data < t.data:
            t.left = self.inserth(t.left, data)
            if self.height(t.left) - self.height(t.right) == 2:
                if data < t.left.data: # Single Rotate
                    t = self.single_rotate_left_to_right(t)
                elif data > t.left.data: # Double Rotate
                    t = self.double_rotate_right_left(t)
            
        elif data > t.data:
            t.right = self.inserth(t.right, data)
            if self.height(t.right) - self.height(t.left) == 2:
                if data > t.right.data:
                    t = self.single_rotate_right_to_left(t)
                elif data < t.right.data:
                    t = self.double_rotate_left_right(t)

        t.height = max(self.height(t.left), self.height(t.right)) + 1
        return t

    def insert(self, data):
        self.root = self.inserth(self.root, data)

    def inorderh(self, t):
        if t:
            self.inorderh(t.left)
            print(t.data, end = ' ')
            self.inorderh(t.right)
    
    def preorderh(self, t):
        if t:
            print(t.data, end = ' ')
            self.preorderh(t.left)
            self.preorderh(t.right)
    
    def postorderh(self, t):
        if t:
            self.postorderh(t.left)
            self.postorderh(t.right)
            print(t.data, end = ' ')

    def levelorderh(self, t):
        if t == None:
            return
        from queue import Queue
        q = Queue()
        q.put(t)
        while(not q.empty()):
            tmp = q.get()
            print(tmp.data, end=' ')
            if tmp.left != None:
                q.put(tmp.left)
            if tmp.right != None:
                q.put(tmp.right)

    def inorder(self):
        self.inorderh(self.root)
        print()

    def preorder(self):
        self.preorderh(self.root)
        print()

    def postorder(self):
        self.postorderh(self.root)
        print()
    
    def levelorder(self):
        self.levelorderh(self.root)
        print()

t = AVLTree()
print(type(t))
print(t)
for item in [77, 33, 55, 99, 44, 11]:
    t.insert(item)
t.inorder()
t.preorder()
t.postorder()
t.levelorder()
print(t.height(t.root))
print(t.root.data)
```

```python
# Maximum Contiguous Sum
def mcs(a):
    m = [0 for i in range(len(a))]
    m[0] = 0 if a[0] < 0 else a[0]
    for i in range(1,len(a)):
        if m[i-1] + a[i] > 0:
            m[i] = m[i-1] + a[i]
        else:
            m[i] = 0
    
    maxi = float("-inf")
    for i in range(0, len(a)):
        if m[i] > maxi:
            maxi = m[i]
    return maxi

print(mcs([-2,11,-4,13,-5,2]))
print(mcs([1,-3,4,-2,-1,6]))
```

```python
# Maximum Contiguous Sum
def mcs(a):
    maxi = 0
    tempsum = 0
    for i in range(0,len(a)):
        if tempsum + a[i] < 0:
            tempsum = 0
            continue
        else:
            tempsum += a[i]
            if tempsum > maxi:
                maxi = tempsum

    return maxi

print(mcs([-2,11,-4,13,-5,2]))
print(mcs([1,-3,4,-2,-1,6]))
```

```python
# Merge Sort
def merge_sort(a, low, high):
    n = high - low + 1
    if n <= 1:
        return
    mid = (low + high) // 2
    merge_sort(a, low, mid)
    merge_sort(a, mid+1, high)
    i = low
    j = mid + 1
    k = 0
    tmp = [0 for _ in range(n)]
    while i <= mid and j <= high:
        if a[i] < a[j]:
            tmp[k] = a[i]
            i += 1
        else:
            tmp[k] = a[j]
            j += 1
        k += 1

    while i <= mid:
        tmp[k] = a[i]
        i += 1
        k += 1
    
    while j <= high:
        tmp[k] = a[j]
        j += 1
        k += 1

    #print(tmp)
    # Write them back to a
    l = n - 1
    z = high
    #print(l)
    #print(z)
    while l >= 0:
        a[z] = tmp[l]
        l -= 1
        z -= 1

    #print(a)

a = [77,33,55,22,44,11,99,-99,100]
merge_sort(a, 0, len(a) - 1)
print(a)
```

```python
class mydict(object):
    HSIZE = 7
    def __init__(self):
        self.table = [list() for _ in range(mydict.HSIZE)]

    def hash_val(self, key):
        import hashlib
        h = hashlib.sha256(key.encode('ascii'))
        return int(h.hexdigest(), base=16) % mydict.HSIZE

    def find(self, key):
        hval = self.hash_val(key)
        lis = self.table[hval]
        for index, item in enumerate(lis):
            if item[0] == key:
                return (hval, index)
        return (hval, None)

    def __setitem__(self, key, value):
        hval, index = self.find(key)
        if index == None:
            self.table[hval].append([key, value])
        else:
            self.table[hval][index][1] = value

    def __getitem__(self, key):
        hval, index = self.find(key)
        if index == None:
            return None
        else:
            return self.table[hval][index][1]

d1 = mydict()
print(type(d1))
d1['ganesh'] = 103
print(d1['ganesh'])
d1['sas'] = 100
d1['sudha'] = 101
print(d1['sas'])
d1['sas'] = 99
print(d1['sas'])
print(d1['sa'])
d1['asdf'] = 'qwer'
d1['asdf'] = 'xyz'
print(d1['asdf'])
print(d1['sudha'])
```

```python
#omgamganapathayenamaha
#Disjoint-Set
class disjoint_set():
    def __init__(self, n, vertices):
        self.n = n
        self.arr = dict()
        for v in vertices:
            self.arr[v] = -1
    
    def find(self, v):
        tmp = v
        while(type(self.arr[tmp]) is str):
            tmp = self.arr[tmp]
            #print(tmp)
        return tmp
    
    # Union by Size
    def union(self, v1, v2):
        fv1 = self.find(v1)
        fv2 = self.find(v2)
        #print(fv1)
        #print(fv2)
        if fv1 == fv2:
            return
        if self.arr[fv1] <= self.arr[fv2]:
            self.arr[fv1] += self.arr[fv2]
            self.arr[fv2] = fv1
        else:
            self.arr[fv2] += self.arr[fv1]
            self.arr[fv1] = fv2

ds = disjoint_set(5, ['A', 'B', 'C', 'D', 'E'])
print(ds.n)
print(ds.arr)
print(ds.find('C'))
ds.union('C', 'D')
print(ds.arr)
ds.union('A','C')
print(ds.arr)
print(ds.find('A'))
ds.union('B', 'E')
print(ds.arr)
ds.union('E', 'A')
print(ds.arr)
ds.union('A', 'B')
print(ds.arr)
```

```python
#omgamganapathayenamaha
#omgamganapathayenamaha
#omgamganapathayenamaha
#omgamganapathayenamaha
#Hash - Open addressing / Closed Hashing
HTABLE_SIZE = 7
LOAD_FACTOR = .7
class hashe():
    def __init__(self):
        self.table_size = HTABLE_SIZE
        self.table = [float("-inf") for _ in range(self.table_size)]
        self.cnt = 0

    def get_hashval(self, key):
        import hashlib
        h = hashlib.sha256(key.encode('ascii'))
        return int(h.hexdigest(), base = 16) % self.table_size

    def get_next_hashval(self, hashval):
        hashval = (hashval + 1) % self.table_size
        return hashval

    def is_prime(self, n):
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return False
        return True

    def next_prime(self, size):
        n = size
        while True:
            n += 1
            if self.is_prime(n):
                break
        print('Next Prime is ' + str(n))
        return n

    def rehash(self):
        old_table = self.table.copy()
        self.cnt = 0
        self.table_size = self.next_prime(self.table_size)
        self.table = [float("-inf") for _ in range(self.table_size)]
        for item in old_table:
            if item != float("-inf"):
                self.__setitem__(item[1], item[0])

    def __setitem__(self, key, val):
        key = str(key)
        hashval = self.get_hashval(key)
        if self.table[hashval] == float("-inf"):
            self.table[hashval] = (val, key)
        else:
            v, k = self.table[hashval]
            while v != float("-inf") and k != key:
                hashval = self.get_next_hashval(hashval)
                if self.table[hashval] == float("-inf"):
                    v = self.table[hashval]
                    k = None
                else:
                    v, k = self.table[hashval]
            self.table[hashval] = (val, key)
        self.cnt += 1
        if (self.cnt / self.table_size) > LOAD_FACTOR:
            #print((self.cnt, key, val))
            self.rehash()

    def __getitem__(self, key):
        key = str(key)
        hashval = self.get_hashval(key)
        if self.table[hashval] == float("-inf"): return None
        v, k = self.table[hashval]
        while v != float("-inf") and k != key:
            hashval = self.get_next_hashval(hashval)
            if self.table[hashval] == float("-inf"):
                v = self.table[hashval]
                k = None
            else:
                v, k = self.table[hashval]
        if k == key:
            return v
        if v == float("-inf"):
            return None

    def __delitem__(self, key):
        key = str(key)
        hashval = self.get_hashval(key)
        if self.table[hashval] == float("-inf"):
            return
        v, k = self.table[hashval]
        while v != float("-inf") and k != key:
            hashval = self.get_next_hashval(hashval)
            if self.table[hashval] == float("-inf"):
                v = self.table[hashval]
                k = None
            else:
                v, k = self.table[hashval]
        if k == key:
            self.table[hashval] = float("-inf")
            self.cnt -= 1

h = hashe()
print(type(h))
print(h)
for item in range(5):
    h[item] = item ** 2
    print(h.table)
print(h[3])
print(h[33])
print(h[2])
h[2] = 44
print(h[3])
print(h[32])
print(h[2])
print(h[0])
h[0] = 100
print(h[0])
print(h[4])
print(h.table)
del h[4]
del h[2]
print(h.table)
h[44] = 444
print(h.table)
print(h[44])
print(h[3])
```
